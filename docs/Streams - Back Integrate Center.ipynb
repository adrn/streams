```python
# Read in data from Kathryn's simulation
sgr_cen = SgrCen()
```

## Integrate SGR_CEN forward and check against file

#### Start by defining the potential with the known parameters:

```python
disk_potential = MiyamotoNagaiPotential(M=1E11*u.M_sun,
                                        a=6.5,
                                        b=0.26)
bulge_potential = HernquistPotential(M=3.4E10*u.M_sun,
                                     c=0.7)
halo_potential = LogarithmicPotentialLJ(v_halo=(121.858*u.km/u.s).to(u.kpc/u.Myr).value,
                                        q1=1.38,
                                        q2=1.0,
                                        qz=1.36,
                                        phi=1.692969,
                                        c=12.)
galaxy_potential = disk_potential + bulge_potential + halo_potential
```

```python
grid = np.arange(-20, 20, 0.1)
galaxy_potential.plot(grid,grid,grid)
```

#### Now we can read in the initial conditions from SGR_CEN, perform a forward integration, and compare to the true values from the data file.

```python
# Read initial conditions / timesteps from data
initial_position = [sgr_cen.data["x"][0], sgr_cen.data["y"][0], sgr_cen.data["z"][0]]
initial_velocity = [sgr_cen.data["vx"][0], sgr_cen.data["vy"][0], sgr_cen.data["vz"][0]]
t1 = min(sgr_cen.data["t"])
t2 = max(sgr_cen.data["t"])

# Integrate the satellite center forward in the potential
ts_forward, xs_forward, vs_forward = leapfrog(galaxy_potential.acceleration_at, 
                                              initial_position, 
                                              initial_velocity, 
                                              t1, t2, sgr_cen.data["dt"][0])
```

For each coordinate, compare position vs. time against timesteps in data file

```python
fig, axes = subplots(1, 3, figsize=(18,8))

dims = ["x", "y", "z"]
for ii, ax in enumerate(axes):
    ax.plot(ts_forward, xs_forward[:,0,ii], color="r", alpha=0.75)
    ax.plot(sgr_cen.data["t"], sgr_cen.data[dims[ii]], color="b", alpha=0.75)
    
    if ii > 0: ax.set_yticklabels([])
    if ii == 0: ax.set_ylabel("kpc")
    ax.set_xlabel("Myr")
    ax.set_title(dims[ii].upper())
    ax.set_xlim(0, 7000)
    ax.set_ylim(-60, 60)

fig.subplots_adjust(wspace=0.09)
```

As a check of the integrator, next I compute the energies for this run (and especially $\Delta E/E$)

```python
# Compute energies for this run
energies_forward = galaxy_potential.energy_at(xs_forward, vs_forward)
delta_E_forward = (energies_forward - energies_forward[0]) / energies_forward[0]

figure(figsize=(16,6))
semilogy(ts_forward, np.fabs(delta_E_forward), color='k', alpha=0.75)
xlabel("Myr")
ylabel(r"$\Delta E / E$")
```
