Branimir's Absolute Magnitude suggestion
================================
I usually use the following equation from Chaboyer (1999) to calculate the absolute magnitude

M = 0.23*[Fe/H] + 0.93

and use [Fe/H] = -1.5 dex for all RR Lyrae stars (median halo metallicity from Ivezic et al. 2008). You may wish to use a different metallicity if you know the metallicity of some stream, but usually this does not matter much.

```python
Fe_H = -1.5
distance_mod = linear_data["magAvg"] - (0.23*Fe_H + 0.93)
distance_kpc = 10**(distance_mod/5. + 1.) / 1000.
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111, projection="aitoff")
ax.scatter(np.radians(180.-ra), np.radians(dec), c=distance_mod, s=1, alpha=0.4)
ax.set_xticklabels([330, 300, 270, 240, 210, 180, 150, 120, 90, 60, 30])
plt.show()
```

Custom Hammer Projection:

![](http://upload.wikimedia.org/math/a/8/0/a8003e2bb4748af207d8025f15876945.png)

![](http://upload.wikimedia.org/math/0/2/3/02354ba061d5eb5a710ed7c51f2b755d.png)

```python
this_ra = ra - 180.
x = 2*np.sqrt(2)*np.cos(np.radians(dec))*np.sin(np.radians(this_ra/2.)) / np.sqrt(1 + np.cos(np.radians(dec))*np.cos(np.radians(this_ra/2.)))
y = np.sqrt(2)*np.sin(np.radians(dec)) / np.sqrt(1 + np.cos(np.radians(dec))*np.cos(np.radians(this_ra/2.)))

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111)
sca = ax.scatter(np.degrees(x)+180., np.degrees(y), c=distance_kpc, s=10, alpha=0.8, edgecolor='none', cmap=cm.Spectral)
ax.set_xlim(220, 120)
fig.colorbar(sca)
ax.set_xlabel("RA")
ax.set_ylabel("Dec")
plt.show()
```

### Next I want to try plotting stars in distance bins, so make this same plot for a bunch of distance bins

```python
distance_step = 4 #kpc
distance_bins = np.arange(0, 40, distance_step)
for bin_left in distance_bins:
    bin_right = bin_left+distance_step
    idx = (distance_kpc >= bin_left) & (distance_kpc < bin_right)
    
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    sca = ax.scatter(np.degrees(x[idx])+180., np.degrees(y[idx]), s=20, alpha=0.75, color='k', edgecolor='none')
    ax.text(140, 70, "{0}-{1} kpc".format(bin_left, bin_right))
    ax.set_xlim(220, 120)
    ax.set_ylim(-40, 80)
    ax.set_xlabel("RA")
    ax.set_ylabel("Dec")
    fig.savefig("/Users/adrian/projects/streams/plots/linear_rrlyr_{0}.png".format(bin_left))
```
