```python
medians = []

Nsamples = sampler.chain.shape[1]
for ii in range(100,Nsamples):
    flatchain = np.vstack(sampler.chain[:,:ii])
    medians.append(list(np.median(flatchain, axis=0)))
medians = np.array(medians)
```

```python
figsize(12,8)
ranges = [(1.3,1.4), (1.3,1.4), (1.65, 1.7), (0.1, 0.15)]
for ii in range(4):
    subplot(4,1,ii+1)
    plot(medians[:,ii], marker=None)
    ylim(ranges[ii])
```
