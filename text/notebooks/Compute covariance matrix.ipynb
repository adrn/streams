```python
def apw_cov1(arr):
    cov = np.zeros((arr.shape[0],arr.shape[0]))
    b = arr - np.mean(arr,axis=1)[:,np.newaxis]
    for ii in range(arr.shape[1]):
        cov += np.outer(b[:,ii],b[:,ii].T)
    cov /= arr.shape[1]
    return cov
```

```python
def objective(potential, s_orbit, p_orbits):
    r_tide = potential._tidal_radius(m=s_orbit._m,
                                     r=s_orbit._r)[:,:,np.newaxis]
    v_esc = potential._escape_velocity(m=s_orbit._m,
                                       r_tide=r_tide)
    
    Ntimesteps,Nparticles,Ndim = p_orbits._r.shape
    
    R,V = relative_normalized_coordinates(potential, p_orbits, s_orbit) 
    D_ps = np.sqrt(np.sum(R**2, axis=-1) + np.sum(V**2, axis=-1))
    
    # Find the index of the time of the minimum D_ps for each particle
    min_time_idx = D_ps.argmin(axis=0)
    
    cov = np.zeros((6,6))
    r_ = p_orbits._r - s_orbit._r
    v_ = p_orbits._v - s_orbit._v
    b = np.vstack((r_.T, v_.T)).T

    for ii in range(Nparticles):
        idx = min_time_idx[ii]
        r_t = np.squeeze(r_tide[idx])
        v_e = np.squeeze(v_esc[idx])
        cov += np.outer(b[idx,ii],b[idx,ii].T)
    cov /= Nparticles
    
    eig_vals = np.linalg.eigvals(cov) / np.array([r_t**2.]*3+[0.012**2.]*3)
    return np.sum(eig_vals)
```

```python
def ln_likelihood(p, param_names, particles, satellite, t1, t2, resolution):
    """ Evaluate the likelihood function for a given set of halo 
        parameters.
    """
    halo_params = dict(zip(param_names, p))
    
    # LawMajewski2010 contains a disk, bulge, and logarithmic halo 
    lm10 = LawMajewski2010(**halo_params)
    v_halo = lm10["halo"]._parameters["v_halo"]
    
    integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
    
    # not adaptive: s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
    s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
    
    #s_orbit,p_orbits = integrator.run(timestep_func=timestep,
    #                                  timestep_args=(),
    #                                  resolution=resolution,
    #                                  t1=t1, t2=t2)
    
    return -objective(lm10, s_orbit, p_orbits)

def ln_posterior(p, *args):
    param_names, particles, satellite, t1, t2, resolution = args
    return ln_prior(p, param_names) + ln_likelihood(p, *args)
```

```python
Nwalkers = 8
params = ['v_halo','phi']

p0 = np.zeros((Nwalkers,len(params)))
for ii,p_name in enumerate(params):
    this_p = np.random.uniform(true_params[p_name]*0.8, 
                               true_params[p_name]*1.2,
                               size=Nwalkers)
    p0[:,ii] = this_p
p0
```

```python
sampler = emcee.EnsembleSampler(nwalkers=Nwalkers, 
                                dim=p0.shape[1], 
                                lnpostfn=ln_posterior, 
                                args=(params, err_particles, satellite, t1, t2, resolution))

old_sampler = emcee.EnsembleSampler(nwalkers=Nwalkers, 
                                dim=p0.shape[1], 
                                lnpostfn=old_ln_posterior, 
                                args=(params, err_particles, satellite, t1, t2, resolution))
```

```python
pos, prob, state = sampler.run_mcmc(p0, 50)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 100)

pos, prob, state = old_sampler.run_mcmc(p0, 50)
old_sampler.reset()
pos, prob, state = old_sampler.run_mcmc(pos, 100)
```
