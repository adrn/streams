```python
import astropy.coordinates as coord
from astropy.table import vstack, Table, Column
import astropy.units as u
from astropy.io import ascii

from scipy.optimize import leastsq

from streams.coordinates.sgr import distance_to_sgr_plane, SgrCoordinates
from streams.io.lm10 import particle_table, particles_today
from streams.io.catalogs import read_stripe82, read_quest
```

```python
stripe82 = read_stripe82()
quest = read_quest()
lm10 = particle_table(expr="(Pcol>-1) & (Pcol<6) & (abs(Lmflag)==1)")
```

```python
data = vstack([stripe82, quest])

stripe82_idx = np.arange(len(data)) < len(stripe82)
quest_idx = np.logical_not(stripe82_idx)

sgr_plane_distance = []
L,B = [], []
l,b = [], []
for star in data:
    d = distance_to_sgr_plane(star['ra'],star['dec'],star['dist'])
    sgr_plane_distance.append(d)
    
    icrs = coord.ICRSCoordinates(star['ra'],star['dec'],unit=(u.deg,u.deg))
    sgr = icrs.transform_to(SgrCoordinates)
    L.append(sgr.Lambda.radians)
    B.append(sgr.Beta.radians)
    l.append(icrs.galactic.l.degrees)
    b.append(icrs.galactic.b.degrees)

Lambda = Column(L, name='Lambda')
Beta = Column(B, name='Beta')
l = Column(l, name='l')
b = Column(b, name='b')
data.add_column(Lambda)
data.add_column(Beta)
data.add_column(l)
data.add_column(b)
```

```python
# only select stars with a distance from the Sgr plane < 10 kpc
idx = (np.fabs(sgr_plane_distance) < 10.) & (data['dist'] < 70)
selected_data = data[idx]
stripe82_idx = stripe82_idx[idx]
quest_idx = quest_idx[idx]
```

# Selection from Nearby Wraps

Now I'll try to do some selection:

* Select only stars in one of the nearby wraps
* Fit a 3D curve to it
* Find all Stripe 82 stars within some distance from the curve

```python
def plot_wrap(wrap):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='polar')
    ax.scatter(wrap['Lambda']*np.pi/180., wrap['dist'],
               edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
    ax.set_ylim(0,70)
```

```python
south_wrap = particle_table(expr="(Pcol>-1) & (Lmflag==1) & (Pcol<7) & "
                                 "(Lambda < 270) & (Lambda > 45)")
north_wrap = particle_table(expr="(Pcol>-1) & (Lmflag==-1) & (Pcol<7) & "
                                 "(Lambda > 180) & (Lambda < 360)")
```

```python
plot_wrap(south_wrap)
plot_wrap(north_wrap)
```

Fit each wrap with a fourier series in r,Lambda,Beta

```python
def model(p, l):
    return p[0] + p[1]*np.sin(l) + p[2]*np.cos(l) + p[3]*np.sin(l)*np.cos(l) +\
           p[4]*np.sin(l)**2 + p[5]*np.cos(l)**2

def err(p, l, r):
    return r - model(p, l)

def fit_wrap(wrap):
    Lambda = wrap['Lambda']*np.pi/180.
    i = np.argsort(Lambda)
    Lambda = Lambda[i]
    p_opt, ier = leastsq(err, np.zeros(6), args=(Lambda,wrap['dist'][i]))
    
    return p_opt
```

```python
fig = plt.figure(figsize=(14,8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

p_opt = fit_wrap(north_wrap)
ax1.scatter(north_wrap['Lambda'], north_wrap['dist'],
            edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
ax1.plot(north_wrap['Lambda'], model(p_opt, north_wrap['Lambda']*np.pi/180.), \
         linestyle='-', marker=None, lw=2)
ax2.scatter(north_wrap['Beta'], north_wrap['dist'],
            edgecolor='none', color='k', marker='o', s=8, alpha=0.25)

p_opt = fit_wrap(south_wrap)
ax1.scatter(south_wrap['Lambda'], south_wrap['dist'],
            edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
ax1.plot(south_wrap['Lambda'], model(p_opt, south_wrap['Lambda']*np.pi/180.), \
         linestyle='-', marker=None, lw=2)
#ax2.scatter(south_wrap['Beta'], south_wrap['dist'],
#            edgecolor='none', color='k', marker='o', s=8, alpha=0.25)

ax1.scatter(selected_data['Lambda']*180/np.pi, selected_data['dist'], 
            edgecolor='none', color='r', marker='o', s=8, alpha=0.75)
ax2.scatter(selected_data['Beta']*180/np.pi, selected_data['dist'], 
            edgecolor='none', color='r', marker='o', s=8, alpha=0.75)
ax1.set_xlabel("$\Lambda$")
ax1.set_ylabel("Distance [kpc]")
ax2.set_xlabel("$B$")
```

Now compute the minimum distance to the fit line, and color the points based on this distance

```python
def plot_flat_wrap(ax, wrap, p_opt):
    wrap.sort('Lambda')
    ax.scatter(wrap['Lambda'], wrap['dist'],
               edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
    ax.plot(wrap['Lambda'], model(p_opt, wrap['Lambda']*np.pi/180.), \
            linestyle='-', marker=None, lw=2)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

p_opt_n = fit_wrap(north_wrap)
plot_flat_wrap(ax, north_wrap, p_opt_n)

p_opt_s = fit_wrap(south_wrap)
plot_flat_wrap(ax, south_wrap, p_opt_s)

ax.scatter(selected_data['Lambda']*180/np.pi, selected_data['dist'], 
           edgecolor='none', color='r', marker='o', s=8, alpha=0.75)
ax.set_xlabel("$\Lambda$")
ax.set_ylabel("Distance [kpc]")
```

Can also do it by binning and averaging in $\Lambda$, $B$

```python
from scipy.interpolate import interp1d
```

```python
def wrap_magic(wrap, Nbins=100):
    lambda_bins = np.linspace(wrap['Lambda'].min(), wrap['Lambda'].max(), Nbins)
    
    beta_per_bin = np.zeros_like(lambda_bins)
    dist_per_bin = np.zeros_like(lambda_bins)
    for ii,binL in enumerate(lambda_bins):
        if ii == Nbins-1: break
        
        binR = lambda_bins[ii+1]
        ps = wrap[(wrap['Lambda'] > binL) & (wrap['Lambda'] < binR)]
        beta_per_bin[ii] = np.median(ps['Beta'])
        dist_per_bin[ii] = np.median(ps['dist'])
    
    beta_func = interp1d(lambda_bins, beta_per_bin, kind='cubic')
    dist_func = interp1d(lambda_bins, dist_per_bin, kind='cubic')
    
    return beta_func, dist_func
```

```python
N_b_func, N_d_func = wrap_magic(north_wrap, Nbins=50)
S_b_func, S_d_func = wrap_magic(south_wrap, Nbins=50)
```

```python
fig = plt.figure(figsize=(6,12))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.scatter(north_wrap['Lambda'], north_wrap['dist'], 
           edgecolor='none', color='k', marker='o', s=8, alpha=0.5)
ax1.plot(north_wrap['Lambda'], N_d_func(north_wrap['Lambda']), 
         color='r', linestyle='-', marker=None)
ax2.scatter(north_wrap['Lambda'], north_wrap['Beta'], 
           edgecolor='none', color='k', marker='o', s=8, alpha=0.5)
ax2.plot(north_wrap['Lambda'], N_b_func(north_wrap['Lambda']), 
         color='r', linestyle='-', marker=None)

#ax.scatter(selected_data['Lambda']*180/np.pi, selected_data['dist'], 
#           edgecolor='none', color='r', marker='o', s=8, alpha=0.75)
ax1.set_ylabel("Distance [kpc]")
ax2.set_ylabel("$B$")
ax2.set_xlabel("$\Lambda$")
```

```python
idx = []
for ii,star in enumerate(selected_data):
    L = star['Lambda']*180./np.pi
    B = star['Beta']*180./np.pi
    D = star['dist']
    B_wrap = None
    try:
        B_wrap = N_b_func(L)
        D_wrap = N_d_func(L)
    except ValueError:
        pass
    
    try:
        B_wrap = S_b_func(L)
        D_wrap = S_d_func(L)
    except ValueError:
        pass
    
    if B_wrap != None:
        if abs(B_wrap-B) < 10. and abs(D_wrap-D) < 7.:
            idx.append(ii)

idx = np.array(idx)
```
