```python
def spherical(x):
    try:
        N = len(x)
    except TypeError: # float
        N = 1
    
    xyz = np.zeros((N,3))
    
    phi = np.random.uniform(0., 2*np.pi, size=N)
    theta = np.arccos(np.random.uniform(-1., 1., size=N))
    
    # x, y, and z components of the vector of magnitude x
    xyz[:,0] = x * np.sin(theta) * np.cos(phi)
    xyz[:,1] = x * np.sin(theta) * np.sin(phi)
    xyz[:,2] = x * np.cos(theta)
    
    return xyz
```

```python
def plummer_particles(N=100, M=2.5E8*u.M_sun, a=1*u.kpc, r_0=None, v_0=None):
    """ Sample N stars from an elliptical galaxy of the specified
        mass, assuming a Plummer model 
        
            $\Phi = - G M / \sqrt{r^2 + a^2}
        
        Note:
            For now, assumes all particle masses = 1 M_sun, but could
            in principle sample from some mass distribution.
        
    """
    
    if not r_0:
        r_0 = [0,0,0.]*u.kpc
    
    if not v_0:
        v_0 = [0,0,0.]*u.kpc/u.Myr
    
    # fractional enclosed mass
    m_M = np.random.uniform(size=N)
    
    # sample positions
    radius = a / np.sqrt(m_M**(-2./3) - 1)
    xyz = spherical(radius) + r_0.to(a.unit).value
    
    # sample velocities, based on positions
    v_mags = np.zeros(N)
    for ii in range(N):
        xx = 0.0
        yy = 0.1
        
        while yy > xx*xx*(1.0-xx*xx)**(3.5):
            xx = np.random.uniform()
            yy = np.random.uniform()*0.1
        
        v_mags[ii] = (xx * np.sqrt(2.0*G*M) * (radius[ii]**2 + a**2) ** -0.25).to(u.kpc/u.Myr).value
    
    v_xyz = spherical(v_mags) + v_0.to(u.kpc/u.Myr).value
    x_v = np.hstack((xyz,v_xyz))
    
    return x_v
```

```python
np.random.seed(42)

potential = CompositePotential(unit_system=UnitSystem.galactic())

unit_system=UnitSystem.galactic()
r_0 = [100.,0.,0.]*u.kpc
v_0 = [-0.19,-0.05,0.]*u.kpc/u.Myr

potential['satellite'] = PlummerPotential(unit_system,
                                          m=2.5E8*u.M_sun,
                                          a=1.*u.kpc,
                                          r_0=r_0)
potential['bulge'] = HernquistPotential(unit_system,
                                        m=3.4E10*u.M_sun,
                                        c=0.7*u.kpc)
                           
potential['disk'] = MiyamotoNagaiPotential(unit_system,
                                           m=1.E11*u.M_sun, 
                                           a=6.5*u.kpc,
                                           b=0.26*u.kpc)
potential['halo'] = LogarithmicPotentialLJ(unit_system,
                                           **true_params)
```

```python
x_v = plummer_particles(N=100, M=2.5E8*u.M_sun, a=1.*u.kpc, r_0=r_0, v_0=v_0)
```

```python
potential_nosat = copy.deepcopy(potential)
del potential_nosat['satellite']
```

```python
x0 = x_v[:,:3]
v0 = x_v[:,3:]
p_integrator = leapfrog.LeapfrogIntegrator(potential._acceleration_at, x0, v0)
s_integrator = leapfrog.LeapfrogIntegrator(potential_nosat._acceleration_at, 
                                           potential['satellite'].parameters['r_0'].value, 
                                           v_0.value)
```

```python
dt = 0.1
s_integrator._prime(dt)
p_integrator._prime(dt)
```

```python
def run(Nsteps, dt, SFR=0.1): #solar mass/yr
    N_p = p_integrator.r_im1.shape[0] + dt*Nsteps*(SFR*1000.)

    all_p_r = np.zeros((Nsteps,N_p,3))
    all_p_v = np.zeros((Nsteps,N_p,3))
    all_s_r = np.zeros((Nsteps,) + s_integrator.r_im1.shape)
    all_s_v = np.zeros((Nsteps,) + s_integrator.r_im1.shape)
    
    r_is = s_integrator.r_im1
    t = [0.]
    for ii in range(Nsteps):
        print ii, t[-1], (t[-1] % 1.)
        potential['satellite'].parameters['r_0'] = r_is*u.kpc
        r_ip,v_ip = p_integrator.step(dt)
        r_is,v_is = s_integrator.step(dt)
        
        all_p_r[ii] = r_ip
        all_p_v[ii] = v_ip
        all_s_r[ii] = r_is
        all_s_v[ii] = v_is
        
        if (t[-1] % 1.) == 0.:
            print t[-1]
            x_v = plummer_particles(N=1000*SFR, M=2.5E8*u.M_sun, a=1.*u.kpc, r_0=r_is*u.kpc, v_0=v_is*u.kpc/u.Myr)
            p_integrator.r_im1 = np.vstack((p_integrator.r_im1, x_v[:,:3]))
            p_integrator.v_im1_2 = np.vstack((p_integrator.v_im1_2, x_v[:,3:]))
        
        t.append(t[-1]+dt)
        
    del t[-1]
    
    particles = OrbitCollection(r=all_p_r*u.kpc,
                                v=all_p_v*u.kpc/u.Myr,
                                t=t*u.Myr,
                                m=1.*u.M_sun)
    satellite = OrbitCollection(r=all_s_r*u.kpc,
                                v=all_s_v*u.kpc/u.Myr,
                                t=t*u.Myr,
                                m=potential['satellite'].parameters['m'])
    
    return particles, satellite
```

```python
p,s = run(Nsteps=10000, dt=0.1)
```
