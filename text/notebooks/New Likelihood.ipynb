```python
import astropy.units as u
import triangle

from streams.io.sgr import mass_selector
from streams.potential.lm10 import LawMajewski2010
from streams.integrate.satellite_particles import satellite_particles_integrate
from streams.inference import minimum_distance_matrix
```

```python
particles_today, satellite_today, time = mass_selector("2.5e7")
```

```python
bound_particles = particles_today(N=0, expr="(tub==0)")
v_disp = np.sqrt(np.sum(np.std(bound_particles._v, axis=0)**2))
```

```python
N = 10

np.random.seed(552)
satellite = satellite_today()
particles = particles_today(N=N, expr="(tub!=0)")
t1,t2 = time()

err = (particles._x*0.01)[0]
```

```python
def rv_at_tub(satellite, particles, potential_params=dict()):
    potential = LawMajewski2010(**potential_params)
    N = len(particles)
    
    acc = np.zeros((N+1,3))
    s,p = satellite_particles_integrate(satellite, particles, potential, 
                                        potential_args=(N+1, acc), 
                                        time_spec=dict(t1=t1, t2=t2, dt=-1.))
    
    ####
    #return minimum_distance_matrix(potential, s, p)
    ####   
    
    ntimesteps = p._x.shape[0]
    
    sat_var = np.zeros((ntimesteps,6))
    sat_var[:,:3] = potential._tidal_radius(satellite._m, s._r) * 1.26 # * 0.6933612743506347 * 2.
    sat_var[:,3:] += v_disp # kpc/Myr
    #sat_var[:,3:] = potential._escape_velocity(satellite._m, r_tide=sat_var[:,:3])
    
    tub_rv = np.zeros((N,6))
    for ii,tub in enumerate(particles.tub):
        t_idx = np.argmin((tub - np.arange(t1,t2,-1.))**2)
        tub_rv[ii] = ((p._x[t_idx,ii] - s._x[t_idx,0]) / sat_var[t_idx])**2.
    
    return tub_rv
```

```python
for qz in [1.2, 1.36, 1.48]:
    figure(figsize=(6,6))
    R_V = rv_at_tub(satellite, particles, potential_params=dict(qz=qz))
    RR, VV = np.sqrt(np.sum(R_V[:,:3]**2, axis=1)), np.sqrt(np.sum(R_V[:,3:]**2, axis=1))
    
    loglog(RR, VV, marker='.', linestyle='none', alpha=0.5)
    xlabel("R")
    ylabel("V")
    xlim(1E-1, 1E5)
    ylim(xlim())
    
    x = np.linspace(xlim()[0], xlim()[1], 100)
    loglog(x, x)
```

```python
# Maybe this is crazy, but could add the "true positions" as parameters?

def likelihood(satellite, particles, potential_params=dict()):
    
    potential = LawMajewski2010(**potential_params)
    N = len(particles)
    
    acc = np.zeros((N+1,3))
    s,p = satellite_particles_integrate(satellite, particles, potential, 
                                        potential_args=(N+1, acc), 
                                        time_spec=dict(t1=t1, t2=t2, dt=-1.))
    
    ntimesteps = p._x.shape[0]
    
    sat_var = np.zeros((ntimesteps,6))
    sat_var[:,:3] = potential._tidal_radius(satellite._m, s._r) * 1.26
    sat_var[:,3:] += v_disp # kpc/Myr
    cov = sat_var**2
    
    l = np.zeros((N,))
    for ii in range(N):
        p_x = p._x[:,ii]
        
        #xx = 0.
        #for tt in range(ntimesteps):
        #    xx += (0.004031441804149937 * (np.prod(cov[tt]))**-0.5) * np.exp(-0.5*np.sum((p_x[tt] - s._x[tt,0])**2 / cov[tt]))
        
        pref = (0.004031441804149937 * (np.prod(cov, axis=1))**-0.5)
        yy = -0.5*np.sum((p_x - s._x[:,0])**2 / cov, axis=1)
        xx = np.sum(pref * np.exp(yy))
        
        l[ii] = np.log(xx)
    
    return np.sum(l)
```
