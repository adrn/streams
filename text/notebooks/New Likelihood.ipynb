```python
def rv_at_tub(satellite, particles, potential_params=dict()):
    potential = LawMajewski2010(**potential_params)
    N = len(particles)
    
    acc = np.zeros((N+1,3))
    s,p = satellite_particles_integrate(satellite, particles, potential, 
                                        potential_args=(N+1, acc), 
                                        time_spec=dict(t1=t1, t2=t2, dt=-1.))
    
    ####
    #return minimum_distance_matrix(potential, s, p)
    ####   
    
    ntimesteps = p._x.shape[0]
    
    sat_var = np.zeros((ntimesteps,6))
    sat_var[:,:3] = potential._tidal_radius(satellite._m, s._r) * 1.26 # * 0.6933612743506347 * 2.
    sat_var[:,3:] += v_disp # kpc/Myr
    #sat_var[:,3:] = potential._escape_velocity(satellite._m, r_tide=sat_var[:,:3])
    
    tub_rv = np.zeros((N,6))
    for ii,tub in enumerate(particles.tub):
        t_idx = np.argmin((tub - np.arange(t1,t2,-1.))**2)
        tub_rv[ii] = ((p._x[t_idx,ii] - s._x[t_idx,0]) / sat_var[t_idx])**2.
    
    return tub_rv
```

```python
# Maybe this is crazy, but could add the "true positions" as parameters?

def likelihood(satellite, particles, potential_params=dict()):
    
    potential = LawMajewski2010(**potential_params)
    N = len(particles)
    
    acc = np.zeros((N+1,3))
    s,p = satellite_particles_integrate(satellite, particles, potential, 
                                        potential_args=(N+1, acc), 
                                        time_spec=dict(t1=t1, t2=t2, dt=-1.))
    
    ntimesteps  = p._x.shape[0]
    
    sat_var = np.zeros((ntimesteps,6))
    sat_var[:,:3] = potential._tidal_radius(satellite._m, s._r) * 1.26
    sat_var[:,3:] += v_disp # kpc/Myr
    cov = sat_var**2
    
    l = np.zeros((N,))
    for ii in range(N):
        p_x = p._x[:,ii]
        
        #xx = 0.
        #for tt in range(ntimesteps):
        #    xx += (0.004031441804149937 * (np.prod(cov[tt]))**-0.5) * np.exp(-0.5*np.sum((p_x[tt] - s._x[tt,0])**2 / cov[tt]))
        
        pref = (0.004031441804149937 * (np.prod(cov, axis=1))**-0.5)
        yy = -0.5*np.sum((p_x - s._x[:,0])**2 / cov, axis=1)
        xx = np.sum(pref * np.exp(yy))
        
        l[ii] = np.log(xx)
    
    return np.sum(l)
```

```python
def new_likelihood(p, potential_param_names, satellite, data_particles, data_errors):
    N = len(data_particles)
    Nparams = len(potential_param_names)
    potential_params = dict(zip(potential_param_names, p[:Nparams]))
    potential = LawMajewski2010(**potential_params)
    
    x = np.array(p[Nparams:Nparams+(N*6)]).reshape(N,6)
    p_d_given_x = 1/np.sqrt(np.prod(data_errors**2)) * np.exp(-0.5 * np.sum((x-data_particles._x)**2 / data_errors**2, axis=-1))
    t_idx = p[Nparams+(N*6):]
    
    particles = Particle(x[:,:3]*u.kpc, x[:,3:]*u.kpc/u.Myr, 0.*u.M_sun)
    
    acc = np.zeros((N+1,3))
    s,p = satellite_particles_integrate(satellite, particles, potential, 
                                        potential_args=(N+1, acc), 
                                        time_spec=dict(t1=t1, t2=t2, dt=-1.))
    
    ntimesteps  = p._x.shape[0]
    
    sat_var = np.zeros((ntimesteps,6))
    sat_var[:,:3] = potential._tidal_radius(satellite._m, s._r) * 1.26
    sat_var[:,3:] += v_disp # kpc/Myr
    cov = sat_var**2
    
    l = np.zeros((N,))
    for ii in range(N):
        pref = (0.004031441804149937 * (np.prod(cov[t_idx[ii]])**-0.5))
        yy = -0.5*np.sum((p._x[t_idx[ii],ii] - s._x[t_idx[ii],0])**2 / cov[t_idx[ii]])
        l[ii] = np.log(pref) + yy + np.log(p_d_given_x[ii])
        #xx = np.sum(pref * np.exp(yy))
        #xx *= p_d_given_x[ii]
        #l[ii] = np.log(xx)
    
    return np.sum(l)
```
