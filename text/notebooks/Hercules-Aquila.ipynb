```python
from astropy.coordinates import GalacticCoordinates
from astropy.io import ascii
from streams.io.lm10 import particle_table
from streams.coordinates import SgrCoordinates
```

```python
p_all = particle_table(N=0, expr="(Pcol > -1) & (abs(Lmflag) == 1) & (Pcol < 7)")
```

```python
p = p_all[p_all['dist'] < 40]
```

```python
herc_data = ascii.read("/Users/adrian/Downloads/rv_heraq.dat")
```

```python
Ls = np.zeros(len(herc_data))
Bs = np.zeros(len(herc_data))
for ii,row in enumerate(herc_data):
    gal = GalacticCoordinates(row['l']*u.degree, row['b']*u.degree)
    sgr = gal.transform_to(SgrCoordinates)
    Ls[ii] = sgr.Lambda.degree
    Bs[ii] = sgr.Beta.degree
```

```python
figure(figsize=(10,6))
plot(p['Lambda'], p['Beta'], marker='.', linestyle='none', alpha=0.25)
plot(Ls, Bs, marker='o', linestyle='none')
xlim(40,15)
ylim(-20, 20)
xlabel(r"$\Lambda_\odot$ [deg]", fontsize=26)
ylabel(r"$B_\odot$ [deg]", fontsize=26)
savefig("/Users/adrian/Downloads/Lambda_Beta.pdf")
```

```python
figure(figsize=(10,6))
plot(p['Lambda'], p['vgsr'], marker='.', linestyle='none', alpha=0.25)
plot(Ls, herc_data['vgsr'], marker='o', linestyle='none')
xlim(40,15)
ylim(-400, 400)
xlabel(r"$\Lambda_\odot$ [deg]", fontsize=26)
ylabel(r"v$_{\rm gsr}$ [km/s]", fontsize=26)
savefig("/Users/adrian/Downloads/Lambda_Vgsr.pdf")
```

```python
figure(figsize=(10,6))
plot(p['l'], p['vgsr'], marker='.', linestyle='none', alpha=0.25)
plot(herc_data['l'], herc_data['vgsr'], marker='o', linestyle='none')
xlim(35,0)
ylim(-400, 400)
xlabel(r"$l$ [deg]", fontsize=26)
ylabel(r"v$_{\rm gsr}$ [km/s]", fontsize=26)
savefig("/Users/adrian/Downloads/l_Vgsr.pdf")
```
