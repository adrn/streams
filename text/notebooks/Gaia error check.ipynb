Stellar types / colors from the Gaia tangential velocity error document

```python
from astropy.io import ascii
```

```python
tbl = ascii.read("""# type M_V V_m_I
A0V 0.7 0.00 
A5V 2.0 0.22 
F0V 2.7 0.47 
F5V 3.5 0.64 
G0V 4.4 0.81 
K0V 5.9 1.06 
K5V 7.3 1.62 
KOIII 0.7 1.30 
K5III -0.2 2.10""")
tbl
```

```python
d_grid = np.logspace(np.log10(50), 4, 100)*u.pc

pm_errors = dict()
for star in tbl:
    pm_errors[star['type']] = []
    for d in d_grid:
        V = apparent_magnitude(star['M_V'], d)
        d_par = parallax_error(V, star['V_m_I']).to(u.microarcsecond)
        dmu = (d_par.value * 0.526) * u.microarcsecond / u.yr
        #dmu = proper_motion_error(V, star['V_m_I']).to(u.microarcsecond/u.yr)
        pm_errors[star['type']].append(dmu.value)
    
    pm_errors[star['type']] = pm_errors[star['type']]*u.microarcsecond/u.yr
```

```python
figure(figsize=(12,8))
for stype in tbl['type']:
    xx = pm_errors[stype].to(u.rad/u.s).value / u.s
    xx = (xx.to(u.km/u.kpc/u.s) * d_grid).to(u.km/u.s)
    loglog(d_grid.value, xx.value*20, label=stype)

legend(loc='upper left')
ylim(0.01, 50)
axvline(1000)
axvline(3000)
```
