```python
import os
```

```python
from streams.coordinates import gc_to_hel, hel_to_gc
from streams.io.lm10 import particles_today, satellite_today
from streams.observation import apparent_magnitude
from streams.observation.gaia import rrl_M_V, parallax_error, proper_motion_error
from streams.observation.rrlyrae import rrl_V_minus_I
from streams.potential.lm10 import LawMajewski2010
from streams.integrate import LeapfrogIntegrator
```

```python
np.random.seed(42)
p = particles_today(N=1, expr="(Pcol>0) & (Pcol<8) & (abs(Lmflag)==1) & (dist<50)")
s = satellite_today()
sat_ptcl = particles_today(N=0, expr="Pcol==-1")

# take a single star from LM10, get heliocentric coordinates
x_,y_,z_ = p.r[0]
vx_,vy_,vz_ = p.v[0]
l_,b_,d_,mul_,mub_,vr_ = gc_to_hel(x_,y_,z_,vx_,vy_,vz_)
```

```python
# get magnitude and color for Gaia errors, Fe/H ~ -0.5 for Sgr
M_V, dM_V = rrl_M_V(-0.5)
V = apparent_magnitude(M_V, d_)

dlb = parallax_error(V, rrl_V_minus_I)
dmu = proper_motion_error(V, rrl_V_minus_I)
```

```python
N_samples = 1000

# astrometric error
l = l_ + np.random.normal(0., dlb.value, size=N_samples) * dlb.unit
b = b_ + np.random.normal(0., dlb.value, size=N_samples) * dlb.unit

# distance error
d = d_ + np.random.normal(0., 0.02*d_.value, size=N_samples) * d_.unit

# proper motion error
mul = mul_ + np.random.normal(0., dmu.value, size=N_samples) * dmu.unit
mub = mub_ + np.random.normal(0., dmu.value, size=N_samples) * dmu.unit

# radial velocity error
vr = vr_ + np.random.normal(0., 10., size=N_samples) * u.km/u.s
```

```python
np.std(vr).to(u.km/u.s)
```

```python
print [np.std(x).value, np.std(y).value, np.std(z).value]
print [np.std(vx).value, np.std(vy).value, np.std(vz).value]
print [x_.value, y_.value, z_.value, vx_.value, vy_.value, vz_.value]
```

```python
x,y,z,vx,vy,vz = hel_to_gc(l,b,d,mul,mub,vr)
r_i = np.array(np.vstack((x.to(u.kpc).value, y.to(u.kpc).value, z.to(u.kpc).value)).T)
v_i = np.array(np.vstack((vx.to(u.kpc/u.Myr).value, vy.to(u.kpc/u.Myr).value, vz.to(u.kpc/u.Myr).value)).T)

r_i = np.vstack((np.array([x_.value,y_.value,z_.value]), r_i))
v_i = np.vstack((np.array([vx_.value,vy_.value,vz_.value]), v_i))

r_i = np.vstack((s._r, r_i))
v_i = np.vstack((s._v, v_i))

plchldr = np.zeros((len(r_i), 3))

print len(r_i)
print np.std(np.sqrt(x**2+y**2+z**2)), np.std(np.sqrt(np.sum(v_i**2, axis=-1)))
```

```python
potential = LawMajewski2010()
integrator = LeapfrogIntegrator(potential._acceleration_at, r_i, v_i, args=(len(r_i), plchldr))
t,r,v = integrator.run(dt=-1., Nsteps=10000)
```

## Relative orbits

```python
r_sat = r[:,0][:,np.newaxis]
v_sat = v[:,0][:,np.newaxis]

r_true = r[:,1][:,np.newaxis]
v_true = v[:,1][:,np.newaxis]

r_tide = potential._tidal_radius(2.5E8, r_sat)[:,:,np.newaxis]
v_disp = np.std(np.sqrt(np.sum(sat_ptcl._v**2, axis=-1))) * np.ones_like(r_tide)

r_true.shape, v_true.shape, r_sat.shape, v_sat.shape, r_tide.shape, v_disp.shape
```

```python
N_p = rv_ptcl.shape[1]

rv_true = np.dstack(((r_true - r_sat) / r_tide, (v_true - v_sat) / v_disp))
rv_ptcl = np.dstack(((r[:,2:] - r_sat) / r_tide, (v[:,2:] - v_sat) / v_disp))
print rv_ptcl.shape, rv_true.shape
```

```python
xxx = hist(np.sqrt(np.sum(rv_true[...,:3]**2, axis=-1)) / np.sqrt(np.sum(rv_true[...,3:]**2, axis=-1)), bins=100)
```

```python
D_ps_ptcl = np.sqrt(np.sum(rv_ptcl**2, axis=-1))
D_ps_true = np.sqrt(np.sum(rv_true**2, axis=-1))
print D_ps_ptcl.shape, D_ps_true.shape

diff = (D_ps_ptcl - D_ps_true)/D_ps_true
```
