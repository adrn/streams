```python
def discrete_cmap(N=8):
    """create a colormap with N (N<15) discrete colors and register it"""
    # define individual colors as hex values
    cpool = ['#9E0142', '#D53E4F', '#F46D43', '#FDAE61', '#FEE08B',
             '#E6F598', '#ABDDA4', '#66C2A5', '#3288BD', '#5E4FA2']
    if N == 5:
        cmap3 = col.ListedColormap(cpool[::2], 'nice_spectral')
    else:
        cmap3 = col.ListedColormap(cpool[0:N], 'nice_spectral')
    cm.register_cmap(cmap=cmap3)
```

```python
from astropy.io import fits
from astropy.table import Table, Column
from streams.data.gaia import rr_lyrae_M_V
```

```python
def rr_lyrae_photometric_distance(m_V, fe_h):
    """ Estimate the distance to an RR Lyrae given its apparent V-band
        magnitude and metallicity.
    """
    M_V,xx = rr_lyrae_M_V(fe_h)
    mu = m_V - M_V
    
    d = 10**(mu/5. + 1) * u.pc
    
    return d.to(u.kpc)
```

```python
Fe_H = -1.5
M_abs = 0.23*Fe_H + 0.93
mu = linear['magAvg'] - M_abs
d = (10.**(mu/5. + 1) * u.pc).to(u.kpc)
```

```python
idx = (d < 31.*u.kpc) & (d > 5.*u.kpc)
dist = d[idx]
linear_dist_cut = linear[idx]
print(len(dist), len(linear_dist_cut))
```

```python
keep_idx = []
Lambda, Beta, D = [], [], []
for ii,star in enumerate(linear_dist_cut):
    icrs = coord.ICRSCoordinates(star["ra"], star["dec"], 
                                 unit=(u.degree,u.degree))
    sgr = icrs.transform_to(SgrCoordinates)
    z_sgr = dist[ii].value * np.sin(sgr.Beta.radians)
    #if abs(sgr.Beta.degrees) < 10.:
    if abs(z_sgr) < 2:
        keep_idx.append(ii)
        Lambda.append(sgr.Lambda.degrees)
        Beta.append(sgr.Beta.degrees)
        D.append(dist[ii].value)
```

```python
sgr_linear = linear_dist_cut[np.array(keep_idx)]
sgr_linear.add_column(at.Column(Lambda, name="Lambda"))
sgr_linear.add_column(at.Column(Beta, name="Beta"))
sgr_linear.add_column(at.Column(D, name="dist"))
                      
len(sgr_linear)
```

```python
def catalog_to_xyz(catalog):
    rrlyr = np.zeros((3,len(catalog)))
    for ii,star in enumerate(catalog):
        icrs = ICRSCoordinates(star["ra"], star["dec"], unit=(u.degree,u.degree))
        gal = icrs.transform_to(GalacticCoordinates)
        
        rrlyr[0,ii] = -star['Dist'] * np.cos(gal.l.radians) * np.sin(gal.b.radians + np.pi/2.)
        rrlyr[1,ii] = star['Dist'] * np.sin(gal.l.radians) * np.sin(gal.b.radians + np.pi/2.)
        rrlyr[2,ii] = star['Dist'] * np.cos(gal.b.radians + np.pi/2.)

    return rrlyr
```
