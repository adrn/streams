```python
R,z,Lz = s.symbols('R,z,L_z')
v0,q = s.symbols('v_0,q')

Phi = v0**2/2*s.log(R**2 + (z/q)**2) + Lz**2/(2*R**2)
Phi
```

```python
dPhi_dr = s.diff(Phi, R)
dPhi_dz = s.diff(Phi, z)
dPhi_dr, dPhi_dz
```

```python
Rdot,zdot = s.symbols(r'\dot{R},\dot{z}')
E = s.Symbol('E')
R,z,Lz = s.symbols('R,z,L_z')
v0,q = s.symbols('v_0,q')
eq = 0.5*(Rdot**2 + zdot**2) + 0.5*s.log(R**2 + z**2/q**2) + Lz**2/(2*R**2) - E
```

```python
eq.subs({s.Symbol('R'):0.3, v0:1., q:0.9, Lz:0.2, E:-0.8, s.Symbol('z'):0.1, zdot:0.1})
```

```python
def acceleration(Rz, v_0, q, L_z):
    R = Rz[...,0]
    z = Rz[...,1]
    return -np.array([eval(str(dPhi_dr.as_expr())),
                     eval(str(dPhi_dz.as_expr()))]).T.copy()
```

```python
np.random.seed(13)
N = 2
RR = np.random.uniform(0.1, 0.4, size=N)
zz = np.random.uniform(-0.1, 0.1, N)
Rz = np.array([RR,zz]).T.copy()
acc = acceleration(Rz, 1., 0.9, 0.2)

zdots = np.array([0.1]*N)
Rdots = []
for (RR,zz),Rd in zip(Rz,zdots):
    Rdots.append(float(s.solve(eq.subs({s.Symbol('R'):RR, s.Symbol('z'):zz, v0:1., q:0.9, Lz:0.2, E:-0.8, zdot:Rd}), Rdot)[1]))

Rzdots = np.array([Rdots,zdots]).T
Rzdots
```

```python
integrator = LeapfrogIntegrator(acceleration, 
                                Rz.copy(), Rzdots.copy(),
                                args=(1., 0.9, 0.2))

t,r,v = integrator.run(dt=0.02, Nsteps=1000)
```
