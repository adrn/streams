```python
def pdf(x,y,r_t,V):
    r = np.sqrt(x**2 + y**2)
    return np.exp(- (r-r_t)**2 / (2*V))
```

```python
x = np.linspace(-5,5,100)
y = x.copy()
X,Y = np.meshgrid(x,y)
X = X.ravel()
Y = Y.ravel()
```

```python
shp = (len(x),len(y))

Z = pdf(X, Y, 3., 3.**2)
figsize(5,5)
pcolor(X.reshape(shp), Y.reshape(shp), Z.reshape(shp), 
       cmap=cm.Blues)
xlim(-5,5)
ylim(-5,5)
```

```python
from streams.io.sgr import SgrSimulation
from streams.integrate import LeapfrogIntegrator
from streams.potential.lm10 import LawMajewski2010
```

```python
sgr = SgrSimulation("2.5e8")
```

```python
p = sgr.particles(N=1000, expr="tub!=0")
```

```python
s = sgr.satellite()
```

```python
plot(p._X[:,0],p._X[:,2],marker='.',linestyle='none',alpha=0.1)
```

```python
lm10_potential = LawMajewski2010()
```

```python
X = np.vstack((s._X[...,:3], p._X[...,:3].copy()))
V = np.vstack((s._X[...,3:], p._X[...,3:].copy()))
integrator = LeapfrogIntegrator(lm10_potential._acceleration_at,
                                np.array(X), np.array(V),
                                args=(X.shape[0], np.zeros_like(X)))
ts, rs, vs = integrator.run(t1=sgr.t1, t2=sgr.t2, dt=-1.)
```

```python
s_orbit = np.vstack((rs[:,0][:,np.newaxis].T, vs[:,0][:,np.newaxis].T)).T
p_orbits = np.vstack((rs[:,1:].T, vs[:,1:].T)).T
t_idx = np.array([np.argmin(np.fabs(ts - t)) for t in p.tub])
```

```python
p_x = np.array([p_orbits[jj,ii] for ii,jj in enumerate(t_idx)])
s_x = np.array([s_orbit[jj,0] for jj in t_idx])
```

```python
diff = p_x-s_x
rel_r = np.sqrt(np.sum(diff[:,:3]**2, axis=-1))
rel_v = np.sqrt(np.sum(diff[:,3:]**2, axis=-1))
```

```python
r_tide = lm10_potential._tidal_radius(2.5e8, s_x)
v_esc = lm10_potential._escape_velocity(2.5e8, r_tide)
```

```python
figsize(11,8)
subplot(121)
title("relative velocity distribution")

#bins = np.logspace(-3, 0, 50)
#xscale('log')
bins = np.linspace(-1e0, 1e0, 100)

xxx = hist(rel_v-v_esc.mean(), bins=bins)
axvline(v_esc.mean(), color='r', label='mean escape velocity')
xlabel(r"$V_{i,sat}$")
legend()

subplot(122)
title("relative distance distribution")

#bins = np.logspace(-1, 2, 50)
#xscale('log')
bins = np.linspace(-1e2, 1e2, 100)

xxx = hist(rel_r-r_tide.mean(), bins=bins)
axvline(r_tide.mean(), color='r', label='mean tidal radius')
xlabel(r"$D_{i,sat}$")
legend()
```
