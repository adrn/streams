```python
def gaussian(x, mu, sigma):
    return 1./np.sqrt(2*np.pi*sigma**2) * np.exp(-0.5 * ((x-mu)/sigma)**2)
```

```python
s_orbit = np.vstack((rs[:,0][:,np.newaxis].T, vs[:,0][:,np.newaxis].T)).T
p_orbits = np.vstack((rs[:,1:].T, vs[:,1:].T)).T
t_idx = np.array([np.argmin(np.fabs(ts - t)) for t in p.tub])

p_x = np.array([p_orbits[jj,ii] for ii,jj in enumerate(t_idx)])
s_x = np.array([s_orbit[jj,0] for jj in t_idx])
```

```python
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

```python
from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

```python
N = 5000
data = dict()
for _m in ["2.5e6", "2.5e7", "2.5e8", "2.5e9"]:
    m = float(_m)
    sgr = SgrSimulation(_m)
    p = sgr.particles(N=N, expr="tub!=0")
    s = sgr.satellite()
    
    p_bound = sgr.particles(N=0, expr="tub==0")
    v_disp = np.sqrt(np.var(p_bound["vx"]) + np.var(p_bound["vy"]) + np.var(p_bound["vz"])).value[0]
    
    X = np.vstack((s._X[...,:3], p._X[...,:3].copy()))
    V = np.vstack((s._X[...,3:], p._X[...,3:].copy()))
    integrator = LeapfrogIntegrator(lm10_potential._acceleration_at,
                                    np.array(X), np.array(V),
                                    args=(X.shape[0], np.zeros_like(X)))
    ts, rs, vs = integrator.run(t1=sgr.t1, t2=sgr.t2, dt=-1.)
    
    s_orbit = np.vstack((rs[:,0][:,np.newaxis].T, vs[:,0][:,np.newaxis].T)).T
    p_orbits = np.vstack((rs[:,1:].T, vs[:,1:].T)).T
    t_idx = np.array([np.argmin(np.fabs(ts - t)) for t in p.tub])
    
    p_x = np.array([p_orbits[jj,ii] for ii,jj in enumerate(t_idx)])
    s_x = np.array([s_orbit[jj,0] for jj in t_idx])
    
    E_s = lm10_potential._value_at(s_x[:,:3]) + 0.5*np.sum(s_x[:,3:]**2, axis=1)
    E_p = lm10_potential._value_at(p_x[:,:3]) + 0.5*np.sum(p_x[:,3:]**2, axis=1)
    
    deltaE = np.fabs(E_s-E_p)
    log_deltaE = np.log(deltaE)
    
    data[_m] = dict(log_deltaE=log_deltaE,
                    v_disp=v_disp,
                    s_orbit=s_orbit,
                    p_x=p_x, s_x=s_x)
```

```python
for _m,d in data.items():
    m = float(_m)
    mu,sigma = norm.fit(d['log_deltaE'])
    r_tide = np.squeeze(lm10_potential._tidal_radius(m, d['s_orbit']))*1.3
    
    idx = r_tide.argmin()
    at_sat = np.squeeze(lm10_potential._value_at(d['s_orbit'][idx,0,:3]))
    at_offset1 = np.squeeze(lm10_potential._value_at(d['s_orbit'][idx,0,:3] + r_tide[idx]))
    at_offset2 = np.squeeze(lm10_potential._value_at(d['s_orbit'][idx,0,:3] - r_tide[idx]))
    huh = np.mean([abs(at_offset1 - at_sat), abs(at_offset2 - at_sat)])
    
    figure()
    n,bins,pat = hist(d['log_deltaE'], bins=np.linspace(-8,-3,100), 
                      alpha=0.5, normed=True)
    x = np.linspace(d['log_deltaE'].min(), d['log_deltaE'].max(), 100)
    axvline(np.log(huh))
    plot(x, gaussian(x, mu, sigma))
    
    muE = np.log(huh)
    sigmaE = 0.35
    plot(x, gaussian(x, muE, sigmaE))
    title(_m)
    xlim(-8,-3)
    
    figure()
    muE = np.log(huh)
    sigmaE = 0.35
#     n,bins,pat = hist(np.exp(-0.5*((d['log_deltaE']-muE)/sigmaE)**2)/np.sqrt(2*np.pi*sigmaE**2), 
#                       bins=np.linspace(-2.,2.,50), alpha=0.5, normed=True)
    n,bins,pat = hist((d['log_deltaE']-muE)/sigmaE, 
                      bins=np.linspace(-2.,2.,50), alpha=0.5, normed=True)
    title(_m)
    xlim(-3,3)
    break
```
