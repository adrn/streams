## Select targets for Spitzer from the Catalina Sky Survey

_Note that this is only the RR Lyrae that have radial velocities, as we want._

```python
import astropy.units as u
from astropy.io import ascii
from astropy.table import join, Column
from scipy.interpolate import interp1d
```

```python
from streams.io import add_sgr_coordinates
from streams.coordinates import distance_to_sgr_plane
from streams.io.lm10 import particle_table
```

```python
lm10 = particle_table(N=0, expr="(Pcol>-1) & (Pcol<8) & (abs(Lmflag)==1) & (dist<100)")
```

```python
spatial_data = ascii.read("../../data/catalog/Catalina_all_RRLyr.txt")
velocity_data = ascii.read("../../data/catalog/Catalina_vgsr_RRLyr.txt")
catalina = join(spatial_data, velocity_data, join_type='outer', keys="ID")
catalina.rename_column("RAdeg", "ra")
catalina.rename_column("DEdeg", "dec")
catalina.rename_column("dh", "dist")
```

We add Sgr coordinates to the table of Catalina stars:

```python
catalina = add_sgr_coordinates(catalina)
print catalina.colnames
```

Now we select:

 * Stars < 10 kpc from the orbital plane of Sgr
 * Stars with distances > 15 kpc
 * Only stars with measured radial velocities

```python
print catalina.colnames
```

```python
sgr_catalina = catalina[np.fabs(catalina["Z_sgr"]) < 10.]
sgr_catalina = sgr_catalina[sgr_catalina["dist"] > 10]
sgr_catalina = sgr_catalina[~sgr_catalina["Vgsr"].mask]
msg = "Out of {0} initial catalog stars, only {1} have radial velocities and match the above criteria."
print msg.format(len(catalina), len(sgr_catalina))
```

Now bin on Sgr plane distance

```python
Nbins = 25
Lambda_bins = np.linspace(north_wrap["Lambda"].min(), north_wrap["Lambda"].max(), Nbins)

median_zsgr = []
d_median_zsgr = []
for ii,binL in enumerate(Lambda_bins):
    if ii == Nbins-1: break
        
    binR = Lambda_bins[ii+1]
    idx = (north_wrap["Lambda"] > binL) & (north_wrap["Lambda"] < binR)
    median_zsgr.append(np.median(north_wrap["sgr_plane_dist"][idx]))
    d_median_zsgr.append(np.std(north_wrap["sgr_plane_dist"][idx]))
median_zsgr = np.array(median_zsgr)
d_median_zsgr = np.array(d_median_zsgr)

zsgr_func = interp1d((Lambda_bins[1:]+Lambda_bins[:-1])/2., 
                     median_zsgr, kind='cubic', bounds_error=False)
zsgr_top_scatter_func = interp1d((Lambda_bins[1:]+Lambda_bins[:-1])/2., 
                                 median_zsgr+2*d_median_zsgr, kind='cubic', bounds_error=False)
zsgr_bot_scatter_func = interp1d((Lambda_bins[1:]+Lambda_bins[:-1])/2., 
                                 median_zsgr-2*d_median_zsgr, kind='cubic', bounds_error=False)
```

```python
def spitzer_integration_time(D):
    return 0.6*(D/10.)**2*u.hour
```

```python
total_sgr_time = sum(spitzer_integration_time(final_sample['dist']))
total_sgr_time
```

```python
def haversine_sphere_dist(lon1, lat1, lon2, lat2):
    from numpy import arcsin, sin, cos

    sdlat = sin((lat2.radian - lat1.radian) / 2)
    sdlon = sin((lon2.radian - lon1.radian) / 2)
    coslats = cos(lat1.radian) * cos(lat2.radian)

    return (2 * arcsin((sdlat ** 2 + coslats * sdlon ** 2) ** 0.5)*u.radian)
```

```python
for star in final_sample:
    others = final_sample[final_sample['ID'] != star['ID']]
    d = haversine_sphere_dist(np.array(star['ra'])*u.deg, np.array(star['dec'])*u.deg,
                              np.array(others['ra'])*u.deg, np.array(others['dec'])*u.deg)
    if min(d.to(u.arcmin).value) < 5.2:
        print "found one!"
```
