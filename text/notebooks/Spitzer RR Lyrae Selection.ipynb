Now we select:

 * Stars < 20 kpc from the orbital plane of Sgr
 * Stars with galactocentric distances > 15 kpc
 * Only stars with measured radial velocities

```python
def tbl_to_xyz(tbl):
    g = coord.ICRS(np.array(tbl["ra"])*u.deg, np.array(tbl["dec"])*u.deg,
                   distance=np.array(tbl["dist"])*u.kpc).galactic
    return g.x-8*u.kpc, g.y, g.z

def tbl_to_gc_dist(tbl):
    x,y,z = tbl_to_xyz(tbl)
    return np.sqrt(x**2 + y**2 + z**2)
```

### Create templates on Lambda vs. Distance for leading/trailing

```python
dist_funcs = {}
median_dists = {}
Lambda_bins = {}
dist_scatters = {}
for lmflag in [-1,1]:
    Nbins = 30
    
    wrap = lm10[lm10['Lmflag'] == lmflag]
    dist = tbl_to_gc_dist(wrap).value
    bins = np.linspace(wrap["Lambda"].min(), wrap["Lambda"].max(), Nbins)
    
    median_dist = []
    dist_scatter = []
    for ii,binL in enumerate(bins):
        if ii == Nbins-1: break
            
        binR = bins[ii+1]
        idx = (wrap["Lambda"] > binL) & (wrap["Lambda"] < binR)
        median_dist.append(np.median(dist[idx]))
        dist_scatter.append(np.std(dist[idx]))
    
    Lambda_bins[lmflag] = (bins[1:]+bins[:-1])/2.
    median_dists[lmflag] = np.array(median_dist)
    dist_scatters[lmflag] = np.array(dist_scatter)
    dist_funcs[lmflag] = interp1d(Lambda_bins[lmflag], 
                                  median_dists[lmflag], 
                                  kind='cubic', bounds_error=False)
```

## Targets:

```python
css_dist = tbl_to_gc_dist(sgr_catalina).value

# Trailing
lmflag = -1
predicted_dist = dist_funcs[lmflag](sgr_catalina["Lambda"])
dist_trail_idx = ((np.fabs(predicted_dist - css_dist)/predicted_dist) < 0.35) & (sgr_catalina["Lambda"] < 180)

# Leading
lmflag = 1
predicted_dist = dist_funcs[lmflag](sgr_catalina["Lambda"])
dist_lead_idx = ((np.fabs(predicted_dist-css_dist)/predicted_dist) < 0.35)
```

Now do selection with RVs:

```python
rv_funcs = {}
median_rvs = {}
Lambda_bins = {}
_Lambda_bins = {}
rv_scatters = {}
for lmflag in [-1,1]:
    Nbins = 30
    
    wrap = lm10[lm10['Lmflag'] == lmflag]
    bins = np.linspace(wrap["Lambda"].min(), wrap["Lambda"].max(), Nbins)
    
    median_rv = []
    rv_scatter = []
    for ii,binL in enumerate(bins):
        if ii == Nbins-1: break
            
        binR = bins[ii+1]
        idx = (wrap["Lambda"] > binL) & (wrap["Lambda"] < binR)
        median_rv.append(np.median(wrap["vgsr"][idx]))
        rv_scatter.append(np.std(wrap["vgsr"][idx]))
    
    Lambda_bins[lmflag] = (bins[1:]+bins[:-1])/2.
    _Lambda_bins[lmflag] = np.array(bins)
    median_rvs[lmflag] = np.array(median_rv)
    rv_scatters[lmflag] = np.array(rv_scatter)
    rv_funcs[lmflag] = interp1d(Lambda_bins[lmflag], 
                                median_rvs[lmflag], 
                                kind='cubic', bounds_error=False)
```

```python
sig_cut = 1.

# Trailing
lmflag = -1
rv_trail_idx = np.zeros_like(dist_targets["Lambda"]).astype(bool)
for ii in range(Nbins-1):
    lbin = _Lambda_bins[lmflag][ii]
    rbin = _Lambda_bins[lmflag][ii+1]
    
    ix = (dist_targets["Lambda"] >= lbin) | (dist_targets["Lambda"] < rbin)
    pred_rv = rv_funcs[lmflag](dist_targets["Lambda"])
    ix &= (dist_targets["Vgsr"] < (pred_rv+sig_cut*rv_scatters[lmflag][ii])) & \
          (dist_targets["Vgsr"] > (pred_rv-sig_cut*rv_scatters[lmflag][ii]))
    
    rv_trail_idx |= ix
    
#predicted_rv = rv_funcs[lmflag](dist_targets["Lambda"])
#rv_trail_idx = ((np.fabs(predicted_rv - dist_targets["Vgsr"])/np.fabs(predicted_rv)) < 0.25)

# Leading
lmflag = 1
rv_lead_idx = np.zeros_like(dist_targets["Lambda"]).astype(bool)
for ii in range(Nbins-1):
    lbin = _Lambda_bins[lmflag][ii]
    rbin = _Lambda_bins[lmflag][ii+1]
    
    ix = (dist_targets["Lambda"] >= lbin) | (dist_targets["Lambda"] < rbin)
    pred_rv = rv_funcs[lmflag](dist_targets["Lambda"])
    ix &= (dist_targets["Vgsr"] < (pred_rv+sig_cut*rv_scatters[lmflag][ii])) & \
          (dist_targets["Vgsr"] > (pred_rv-sig_cut*rv_scatters[lmflag][ii]))
    
    rv_lead_idx |= ix
    
#predicted_rv = rv_funcs[lmflag](dist_targets["Lambda"])
#rv_lead_idx = ((np.fabs(predicted_rv - dist_targets["Vgsr"])/np.fabs(predicted_rv)) < 0.25)
```

```python
rv_idx = rv_lead_idx | rv_trail_idx
rv_targets = dist_targets[rv_idx]
not_rv_targets = dist_targets[~rv_idx]
```

```python
def haversine_sphere_dist(lon1, lat1, lon2, lat2):
    from numpy import arcsin, sin, cos

    sdlat = sin((lat2.radian - lat1.radian) / 2)
    sdlon = sin((lon2.radian - lon1.radian) / 2)
    coslats = cos(lat1.radian) * cos(lat2.radian)

    return (2 * arcsin((sdlat ** 2 + coslats * sdlon ** 2) ** 0.5)*u.radian)
```

```python
for star in final_sample:
    others = final_sample[final_sample['ID'] != star['ID']]
    d = haversine_sphere_dist(np.array(star['ra'])*u.deg, np.array(star['dec'])*u.deg,
                              np.array(others['ra'])*u.deg, np.array(others['dec'])*u.deg)
    if min(d.to(u.arcmin).value) < 5.2:
        print "found one!"
```
