## Wavelength calibration

```python
import os, sys
from astropy.io import fits
from astropy.modeling.polynomial import Poly1DModel
from scipy.optimize import leastsq
```

```python
# For Hg-Ne lamps
line_list = np.array([5460.735, 5769.598, 5790.663, 5852.488, 5881.895, 
                      5944.834, 5975.534, 6029.997, 6074.338, 6096.163,
                      6128.450, 6143.062, 6163.594, 6217.281, 6266.495,
                      6304.789, 6334.428, 6382.992, 6402.246, 6506.528,
                      6532.882, 6598.953, 6678.276, 6717.043, 6929.467,
                      7032.413, 7173.938, 7245.167])
```

```python
# read in COMP
hdu = fits.open(os.path.join(data_path, "m090213.0047.fit"))[0]
hg_ne = np.sum(hdu.data[:,200:300], axis=1)
hg_ne /= hg_ne.max()
```

```python
plot(hg_ne)
```

```python
degree = 4
scale_model = Poly1DModel(degree)
```

```python
def model(p, x):
    scale_model.parameters = list(p[:degree+1])
    
    sigma = p[degree+1]
    amps = p[degree+2:]
    mus = line_list * scale_model(line_list)
    
    m = 0.
    for ii in range(len(amps)):
        m0 = amps[ii] * np.exp(-0.5 * (x - mus[ii])**2 / (2*sigma**2))
        m += m0
    
    return m
```

```python
def err_func(p, x, flux):
    return flux - model(p,x)
```

```python
#p = [0,0,0,0,-5000.]
p = [0.1,1.,1.,1.,1.]
p += [2.]
p += [1.]*len(line_list)

x = np.arange(1024)
p_opt, ier = leastsq(err_func, x0=p, args=(x, hg_ne), maxfev=1000000)
ier
```

```python
figure(figsize=(12,6))
#plot(x, hg_ne)
plot(x, model(p_opt, x))
```
