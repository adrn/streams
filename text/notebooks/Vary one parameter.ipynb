```python
from streams.io.lm10 import particles_today, satellite_today, time
from streams.inference.lm10 import ln_likelihood, timestep
from streams.potential.lm10 import LawMajewski2010, true_params
from streams.integrate.satellite_particles import SatelliteParticleIntegrator
from streams.inference import minimum_distance_matrix
```

```python
particles = particles_today(N=100, expr="(Pcol>-1)&(abs(Lmflag)==1)")
satellite = satellite_today()
t1,t2 = time()
resolution = 3.
```

```python
def relative_normalized_coordinates(potential, particle_orbits, satellite_orbit):
    # need to add a new axis to normalize each coordinate component
    r_tide = potential._tidal_radius(m=satellite_orbit._m,
                                     r=satellite_orbit._r)[:,:,np.newaxis]
    v_esc = potential._escape_velocity(m=satellite_orbit._m,
                                       r_tide=r_tide)
    
    return (particle_orbits._r - satellite_orbit._r) / (r_tide), \
           (particle_orbits._v - satellite_orbit._v) / v_esc

def minimum_distance_matrix(potential, particle_orbits, satellite_orbit):
    Nparticles = particle_orbits._r.shape[1]
    
    R,V = relative_normalized_coordinates(potential, particle_orbits, satellite_orbit) 
    D_ps = np.sqrt(np.fabs(np.sum(R**2, axis=-1)-1) + np.sum(V**2, axis=-1))
    
    # Find the index of the time of the minimum D_ps for each particle
    min_time_idx = D_ps.argmin(axis=0)
    
    min_ps = np.zeros((Nparticles,6))
    xx = zip(min_time_idx, range(Nparticles))
    for kk in range(Nparticles):
        jj,ii = xx[kk]
        min_ps[ii] = np.append(R[jj,ii], V[jj,ii])
    
    return min_ps
```

```python
def generalized_variance_sum(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix(potential, particle_orbits, satellite_orbit)
    
    cov_matrix = np.cov(min_ps.T)
    # cov_matrix -> (6 x 6) covariance matrix for particles
    w,v = np.linalg.eig(cov_matrix)
    
    return np.sum(w)

def generalized_variance_prod(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix(potential, particle_orbits, satellite_orbit)
    
    cov_matrix = np.cov(min_ps.T)
    # cov_matrix -> (6 x 6) covariance matrix for particles
    w,v = np.linalg.eig(cov_matrix)
    
    return np.prod(w)

def just_variance(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix(potential, particle_orbits, satellite_orbit)
    D_ps = np.sum(min_ps**2, axis=-1)
    return np.var(D_ps)
```

```python
x = np.linspace(0.9, 1.1, 21)
variances = dict()
for param in ['q1','qz','v_halo','phi']:
    if not variances.has_key(param):
        variances[param] = []
        
    stats = x*true_params[param]
    for stat in stats:
        params = true_params.copy()
        params[param] = stat
        lm10 = LawMajewski2010(**params)
        integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
        s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
        #variances[param].append(generalized_variance_sum(lm10, p_orbits, s_orbit))
        variances[param].append(generalized_variance_prod(lm10, p_orbits, s_orbit))
        #variances[param].append(just_variance(lm10, p_orbits, s_orbit))
```
