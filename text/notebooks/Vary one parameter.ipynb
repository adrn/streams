```python
import emcee
from streams.io.lm10 import particles_today, satellite_today, time
from streams.inference.lm10 import ln_prior, ln_likelihood, timestep
from streams.potential.lm10 import LawMajewski2010, true_params
from streams.integrate.satellite_particles import SatelliteParticleIntegrator
from streams.inference import minimum_distance_matrix, relative_normalized_coordinates, \
                              generalized_variance
from streams.observation.gaia import add_uncertainties_to_particles
```

```python
def minimum_distance_matrix2(potential, particle_orbits, satellite_orbit):
    Nparticles = particle_orbits._r.shape[1]
    
    R,V = relative_normalized_coordinates(potential, particle_orbits, satellite_orbit) 
    D_ps = np.sqrt(np.sum(R**2, axis=-1) + np.sum(V**2, axis=-1))
    
    # Find the index of the time of the minimum D_ps for each particle
    min_time_idx = D_ps.argmin(axis=0)
    
    min_ps = np.zeros((Nparticles,6))
    xx = zip(min_time_idx, range(Nparticles))
    for kk in range(Nparticles):
        jj,ii = xx[kk]
        min_ps[ii] = np.append(R[jj,ii], V[jj,ii])
    
    return min_ps

def generalized_variance2(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix2(potential, particle_orbits, satellite_orbit)
    cov_matrix = np.cov(np.fabs(min_ps.T))
    sign,logdet = np.linalg.slogdet(cov_matrix)
    return logdet
```

```python
true_particles = particles_today(N=100, expr="(Pcol>-1) & (abs(Lmflag)==1)")
satellite = satellite_today()
t1,t2 = time()
resolution = 4.
```

```python
particles = add_uncertainties_to_particles(true_particles, 
                                           radial_velocity_error=10.*u.km/u.s,
                                           distance_error_percent=0.02)
```

```python
x = np.linspace(0.8, 1.2, 21)
variances = dict()
variances2 = dict()
for param in ['q1','qz','v_halo','phi']:
    if not variances.has_key(param):
        variances[param] = []
        variances2[param] = []
        
    stats = x*true_params[param]
    for stat in stats:
        params = true_params.copy()
        params[param] = stat
        lm10 = LawMajewski2010(**params)
        integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
        s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
        
        variances[param].append(generalized_variance(lm10, p_orbits, s_orbit))
        variances2[param].append(generalized_variance2(lm10, p_orbits, s_orbit))
```

```python
figure(figsize=(8,8))
for ii,param in enumerate(['q1','qz','v_halo','phi']):
    subplot(2,2,ii+1)
    title(param)
    plot(x,np.array(variances[param]),color='k')
    #plot(x,np.array(variances2[param]),color='r')

subplots_adjust(wspace=0.25, hspace=0.4)
```

```python
def ln_like(p, param_names):
    params = dict(zip(param_names, p))
    lm10 = LawMajewski2010(**params)
    integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
    s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
    var = generalized_variance(lm10, p_orbits, s_orbit)
    return -2**var

def objective(p, param_names):
    return ln_prior(p, param_names) + ln_like(p, param_names)
```

```python
Nwalkers = 6
params = ['q1','v_halo']

p0 = np.zeros((Nwalkers,len(params)))
for ii,p_name in enumerate(params):
    this_p = np.random.uniform(true_params[p_name]*0.8, 
                               true_params[p_name]*1.2,
                               size=Nwalkers)
    p0[:,ii] = this_p
```

```python
sampler = emcee.EnsembleSampler(nwalkers=Nwalkers, 
                                dim=p0.shape[1], 
                                lnpostfn=objective, 
                                args=(params,))
```

```python
pos, prob, state = sampler.run_mcmc(p0, 25)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 25)
```

```python
idx = sampler.flatlnprobability.argmax()
```

```python
flatchain = np.vstack(sampler.chain)

subplot(121)
hist(flatchain[:,0], bins=np.linspace(0.9,1.1,51)*true_params['q1'])
axvline(np.median(flatchain[:,0]), color='b')
axvline(flatchain[idx,0], color='g')
axvline(true_params['q1'], color='r')

subplot(122)
hist(flatchain[:,1], bins=np.linspace(0.9,1.1,51)*true_params['v_halo'])
axvline(np.median(flatchain[:,1]), color='b')
axvline(flatchain[idx,1], color='g')
axvline(true_params['v_halo'].value, color='r')
```
