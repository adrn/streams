```python
from streams.data import lm10_particles, lm10_satellite, lm10_time
from streams.inference.lm10 import ln_likelihood, timestep
from streams.potential.lm10 import LawMajewski2010, true_params
from streams.integrate.satellite_particles import SatelliteParticleIntegrator
from streams.inference import minimum_distance_matrix
```

```python
particles = lm10_particles(N=10, expr="(Pcol>0)&(abs(Lmflag)==1)")
satellite = lm10_satellite()
t1,t2 = lm10_time()
resolution = 3.
```

```python
def generalized_variance_sum(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix(potential, particle_orbits, satellite_orbit)
    
    cov_matrix = np.cov(min_ps.T)
    # cov_matrix -> (6 x 6) covariance matrix for particles
    w,v = np.linalg.eig(cov_matrix)
    
    return np.sum(w)

def generalized_variance_prod(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix(potential, particle_orbits, satellite_orbit)
    
    cov_matrix = np.cov(min_ps.T)
    # cov_matrix -> (6 x 6) covariance matrix for particles
    w,v = np.linalg.eig(cov_matrix)
    
    return np.prod(w)
```

```python
timestep2 = lambda *args,**kwargs: -2.
resolution = 1.
```

```python
variances = dict()
for param in ['q1','qz','v_halo','phi']:
    if not variances.has_key(param):
        variances[param] = []
        
    stats = np.linspace(true_params[param]*0.9,
                        true_params[param]*1.1, 
                        10)

    for stat in stats:
        params = true_params.copy()
        params[param] = stat
        lm10 = LawMajewski2010(**params)
        integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
        s_orbit,p_orbits = integrator.run(timestep_func=timestep2,
                                  timestep_args=(lm10, satellite.m.value),
                                  resolution=resolution,
                                  t1=t1, t2=t2)
        variances[param].append(generalized_variance_prod(lm10, p_orbits, s_orbit))
```

```python
for param in ['q1','qz','v_halo','phi']:
    x = np.linspace(0.9, 1.1, 10)
    semilogy(x,variances[param],label=param)
legend()
```
