```python
from streams.io.lm10 import particles_today, satellite_today, time
from streams.inference.lm10 import ln_likelihood, timestep
from streams.potential.lm10 import LawMajewski2010, true_params
from streams.integrate.satellite_particles import SatelliteParticleIntegrator
from streams.inference import minimum_distance_matrix, relative_normalized_coordinates, \
                              generalized_variance
```

```python
particles = particles_today(N=100, expr="(Pcol>-1)&(abs(Lmflag)==1)")
satellite = satellite_today()
t1,t2 = time()
resolution = 4.
```

```python
def minimum_distance_matrix2(potential, particle_orbits, satellite_orbit):
    Nparticles = particle_orbits._r.shape[1]
    
    R,V = relative_normalized_coordinates(potential, particle_orbits, satellite_orbit) 
    D_R = np.fabs(np.sqrt(np.sum(R**2, axis=-1)) - 1.)
    D_ps = np.sqrt(D_R**2. + np.sum(V**2, axis=-1))
    
    # Find the index of the time of the minimum D_ps for each particle
    min_time_idx = D_ps.argmin(axis=0)
    
    min_ps = np.zeros((Nparticles,6))
    xx = zip(min_time_idx, range(Nparticles))
    for kk in range(Nparticles):
        jj,ii = xx[kk]
        min_ps[ii] = np.append(R[jj,ii], V[jj,ii])
    
    return min_ps

def generalized_variance2(potential, particle_orbits, satellite_orbit):
    min_ps = minimum_distance_matrix2(potential, particle_orbits, satellite_orbit)
    cov_matrix = np.cov(np.fabs(min_ps.T))
    sign,logdet = np.linalg.slogdet(cov_matrix)
    return logdet
```

```python
x = np.linspace(0.9, 1.1, 21)
variances = dict()
variances2 = dict()
for param in ['q1','qz','v_halo','phi']:
    if not variances.has_key(param):
        variances[param] = []
        variances2[param] = []
        
    stats = x*true_params[param]
    for stat in stats:
        params = true_params.copy()
        params[param] = stat
        lm10 = LawMajewski2010(**params)
        integrator = SatelliteParticleIntegrator(lm10, satellite, particles)
        s_orbit,p_orbits = integrator.run(t1=t1, t2=t2, dt=-1.)
        
        variances[param].append(generalized_variance(lm10, p_orbits, s_orbit))
        variances2[param].append(generalized_variance2(lm10, p_orbits, s_orbit))
```

```python
for ii,param in enumerate(['q1','qz','v_halo','phi']):
    subplot(2,2,ii+1)
    title(param)
    plot(x,-np.array(variances[param])**2)
    plot(x,-np.array(variances2[param])**2)

subplots_adjust(wspace=0.25, hspace=0.4)
```
