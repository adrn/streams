```python
import astropy.units as u
from astropy.constants import G
from streams.integrate import LeapfrogIntegrator
from streams.potential import SphericalNFWPotential
from streams.potential.lm10 import LawMajewski2010
```

We start with a spherical NFW potential:

```python
potential = SphericalNFWPotential(units=[u.kpc, u.Myr, u.radian, u.M_sun], 
                                  log_m=np.log(8.E11),
                                  a=12*u.kpc)
#potential.plot(ndim=3, grid=np.linspace(-50,50,100)*u.kpc)
```

We find the circular velocity at 75 kpc, and run a test orbit of a satellite at this radius and velocity:

```python
potential.parameters
```

```python
potential.v_circ(8*u.kpc).to(u.km/u.s)
```

```python
R = 50*u.kpc
vc = potential.v_circ(8*u.kpc).to(u.km/u.s).value[0]

initial_position = np.array([R.value, 0.0, 0.]) # kpc
initial_velocity = (np.array([0., vc, 0.])*u.km/u.s)\
                            .to(u.kpc/u.Myr).value # kpc/Myr

integrator = LeapfrogIntegrator(potential._acceleration_at,
                        initial_position.T, initial_velocity.T)
ts, xs, vs = integrator.run(t1=0., t2=6000., dt=1.)
```

```python
grid = np.linspace(-75, 75., 250)*u.kpc
fig2, axes2 = potential.plot(ndim=3, grid=grid)
axes2[0,0].plot(xs[...,0], xs[...,1], color='w', marker=None, alpha=0.65)
axes2[1,0].plot(xs[...,0], xs[...,2], color='w', marker=None, alpha=0.65)
axes2[1,1].plot(xs[...,1], xs[...,2], color='w', marker=None, alpha=0.65)
```

Now try with a companion at slightly higher velocity, offset in X

```python
sgr = SphericalNFWPotential(units=[u.kpc, u.Myr, u.radian, u.M_sun], 
                            log_m=np.log(2.5E8),
                            a=1*u.kpc)
```

```python
R = 50*u.kpc
dR = 5.*u.kpc
vc = potential.v_circ(8*u.kpc).to(u.kpc/u.Myr).value[0]
dv = sgr.v_circ([dR.value,0,0]*u.kpc).value[0]

initial_position = np.array([[R.value, 0., 0.],
                             [R.value, 0., dR.value]]) # kpc
initial_velocity = np.array([[0., vc, 0.],
                             [0., vc+dv, 2*dv]]) # kpc/Myr

integrator = LeapfrogIntegrator(potential._acceleration_at,
                        initial_position, initial_velocity)
ts, xs, vs = integrator.run(t1=0., t2=1000., dt=1.)
```

```python
grid = np.linspace(-75, 75., 250)*u.kpc
fig2, axes2 = potential.plot(ndim=3, grid=grid)

axes2[0,0].plot(xs[:,0,0], xs[:,0,1], color='w', marker=None, alpha=0.65)
axes2[1,0].plot(xs[:,0,0], xs[:,0,2], color='w', marker=None, alpha=0.65)
axes2[1,1].plot(xs[:,0,1], xs[:,0,2], color='w', marker=None, alpha=0.65)

axes2[0,0].plot(xs[:,1,0], xs[:,1,1], color='r', marker=None, alpha=0.65)
axes2[1,0].plot(xs[:,1,0], xs[:,1,2], color='r', marker=None, alpha=0.65)
axes2[1,1].plot(xs[:,1,1], xs[:,1,2], color='r', marker=None, alpha=0.65)
```

```python
N = 1000
R = 60*u.kpc
dR = 5.*u.kpc
vc = potential.v_circ(8*u.kpc).to(u.kpc/u.Myr).value[0]
dv = sgr.v_circ([dR.value,0,0]*u.kpc).value[0]

sgr_v = np.array([vc/2, vc, 0.])
initial_position = np.random.normal([R.value, 0., 0.], [1., 1., 1.], size=(N,3))
initial_velocity = np.random.normal(sgr_v, ([5.,5.,5.]*u.km/u.s).to(u.kpc/u.Myr).value, size=(N,3))
integrator = LeapfrogIntegrator(potential._acceleration_at,
                                initial_position, initial_velocity)

initial_position2 = np.random.normal([R.value+dR.value, 0., dR.value], [0.5, 0.5, 0.5], size=(N//10,3))
initial_velocity2 = np.random.normal(sgr_v + dv, ([3.,3.,3.]*u.km/u.s).to(u.kpc/u.Myr).value, size=(N//10,3))
integrator2 = LeapfrogIntegrator(potential._acceleration_at,
                                initial_position2, initial_velocity2)

ts, xs, vs = integrator.run(t1=0., t2=6000., dt=1.)
ts, xs2, vs2 = integrator2.run(t1=0., t2=6000., dt=1.)
```

```python
grid = np.linspace(-100, 100, 250)*u.kpc
fig2, axes2 = potential.plot(ndim=3, grid=grid)

axes2[0,0].plot(xs[-1,:,0], xs[-1,:,1], color='w', marker='.', alpha=0.65, linestyle='none')
axes2[1,0].plot(xs[-1,:,0], xs[-1,:,2], color='w', marker='.', alpha=0.65, linestyle='none')
axes2[1,1].plot(xs[-1,:,1], xs[-1,:,2], color='w', marker='.', alpha=0.65, linestyle='none')

axes2[0,0].plot(xs2[-1,:,0], xs2[-1,:,1], color='r', marker='.', alpha=0.65, linestyle='none')
axes2[1,0].plot(xs2[-1,:,0], xs2[-1,:,2], color='r', marker='.', alpha=0.65, linestyle='none')
axes2[1,1].plot(xs2[-1,:,1], xs2[-1,:,2], color='r', marker='.', alpha=0.65, linestyle='none')
```

```python
from streams.io.lm10 import LM10Simulation
```

```python
lm10 = LM10Simulation()
lm10_potential = LawMajewski2010()
```

```python
s = lm10.satellite()
```

```python
xx = s._X[...,:3]
integrator = LeapfrogIntegrator(lm10_potential._acceleration_at,
                                np.array(xx), np.array(s._X[...,3:]),
                                args=(xx.shape[0], np.zeros_like(xx)))
ts, sgr_xs, sgr_vs = integrator.run(t1=0., t2=-6000., dt=-1.)
```

```python
sgr_x = np.squeeze(sgr_xs[-1])
sgr_v = np.squeeze(sgr_vs[-1])
sgr_x, sgr_v
```

```python
N = 1000

initial_position = np.random.normal(sgr_x, [1., 1., 1.], size=(N,3))
initial_velocity = np.random.normal(sgr_v, ([6.,6.,6.]*u.km/u.s).to(u.kpc/u.Myr).value, size=(N,3))
integrator = LeapfrogIntegrator(lm10_potential._acceleration_at,
                                initial_position, initial_velocity,
                                args=(N, np.zeros_like(initial_position)))
ts, xs, vs = integrator.run(t1=0., t2=6000., dt=1.)

dR = 5.*u.kpc
dv = sgr.v_circ([dR.value,0,0]*u.kpc).value[0]
initial_position2 = np.random.normal([R.value+dR.value, 0., dR.value], [0.5, 0.5, 0.5], size=(N//10,3))
initial_velocity2 = np.random.normal(sgr_v + dv, ([3.,3.,3.]*u.km/u.s).to(u.kpc/u.Myr).value, size=(N//10,3))
integrator2 = LeapfrogIntegrator(potential._acceleration_at,
                                initial_position2, initial_velocity2)
ts, xs2, vs2 = integrator2.run(t1=0., t2=6000., dt=1.)
```

```python
grid = np.linspace(-100, 100, 250)*u.kpc
fig2, axes2 = potential.plot(ndim=3, grid=grid)

axes2[0,0].plot(xs[-1,:,0], xs[-1,:,1], color='w', marker='.', alpha=0.65, linestyle='none')
axes2[1,0].plot(xs[-1,:,0], xs[-1,:,2], color='w', marker='.', alpha=0.65, linestyle='none')
axes2[1,1].plot(xs[-1,:,1], xs[-1,:,2], color='w', marker='.', alpha=0.65, linestyle='none')

axes2[0,0].plot(xs2[-1,:,0], xs2[-1,:,1], color='r', marker='.', alpha=0.65, linestyle='none')
axes2[1,0].plot(xs2[-1,:,0], xs2[-1,:,2], color='r', marker='.', alpha=0.65, linestyle='none')
axes2[1,1].plot(xs2[-1,:,1], xs2[-1,:,2], color='r', marker='.', alpha=0.65, linestyle='none')
```
