```python
def uncertainties(p, d_percent_err=0.02, rv_err=20):
    p = p.copy()
    
    # assuming [Fe/H] = -0.5 for Sgr
    M_V, dM_V = rrl_M_V(-0.5)
    
    V = apparent_magnitude(M_V, p['dist']*u.kpc)
    
    p['dist'] += np.random.normal(0., d_percent_err*p['dist'])
    p['vr'] += np.random.normal(0., rv_err, size=len(p))
    
    dmu = proper_motion_error(V, rrl_V_minus_I).to(u.milliarcsecond/u.yr).value
    p['mul'] += np.random.normal(0., dmu, size=len(p))
    p['mub'] += np.random.normal(0., dmu, size=len(p))
    
    cosl = np.cos(np.radians(p['l']))
    cosb = np.cos(np.radians(p['b'])) 
    sinl = np.sin(np.radians(p['l']))
    sinb = np.sin(np.radians(p['b'])) 
    
    p['x'] = p['dist']*cosb*cosl - 8.
    p['y'] = p['dist']*cosb*sinl
    p['z'] = p['dist']*sinb
    
    p['vx'] = (p['vr']*cosb*cosl - p['dist']*p['mul']*sinl - p['dist']*p['mub']*sinb*cosl)
    p['vy'] = (p['vr']*cosb*sinl + p['dist']*p['mul']*cosl - p['dist']*p['mub']*sinb*sinl)
    p['vz'] = (p['vr']*sinb + p['dist']*p['mub']*cosb)
    
    return p
```
