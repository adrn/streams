Subselect 100 particles, add errors

```python
subset = all_particles[np.random.randint(len(all_particles), size=100)]
```

```python
figure(figsize=(7,7))
plot(all_particles['x'], all_particles['z'], marker=',', 
     linestyle='none', alpha=0.3, color='#555555')
plot(subset['x'], subset['z'], marker='+', linestyle='none', markeredgewidth=1,
     alpha=1., color='k', markersize=8, markeredgecolor='k')
xlim(-85.,60)
ylim(-70, 75)
xlabel(r"$X_{\rm GC}$ [kpc]", fontsize=22)
ylabel(r"$Z_{\rm GC}$ [kpc]", fontsize=22)
axes().set_aspect('equal')
```

```python
def uncertainties(p_, d_percent_err=0.02, rv_err=20.):
    p = p_.copy()
    
    # assuming [Fe/H] = -0.5 for Sgr
    M_V, dM_V = rrl_M_V(-0.5)
    V = apparent_magnitude(M_V, p['dist']*u.kpc)
    
    p['dist'] += np.random.normal(0., d_percent_err*p['dist'])
    p['vr'] += np.random.normal(0., rv_err, size=len(p))
    
    dmu = proper_motion_error(V, rrl_V_minus_I).to(u.milliarcsecond/u.yr).value
    p['mul'] += np.random.normal(0., dmu, size=len(p))
    p['mub'] += np.random.normal(0., dmu, size=len(p))
    
    mul = (p['mul']*u.milliarcsecond/u.yr).to(u.rad/u.s).value
    mub = (p['mub']*u.milliarcsecond/u.yr).to(u.rad/u.s).value
    
    cosl = np.cos(np.radians(p['l']))
    cosb = np.cos(np.radians(p['b'])) 
    sinl = np.sin(np.radians(p['l']))
    sinb = np.sin(np.radians(p['b'])) 
    
    p['x'] = p['dist']*cosb*cosl - 8.
    p['y'] = p['dist']*cosb*sinl
    p['z'] = p['dist']*sinb
    
    # compute tangential velocities
    d = (p['dist']*u.kpc).to(u.km).value
    v_l = d*mul
    v_b = d*mub
    
    p['vx'] = (p['vr']*cosb*cosl - v_l*sinl - v_b*sinb*cosl) + 9.438
    p['vy'] = (p['vr']*cosb*sinl + v_l*cosl - v_b*sinb*sinl) + 220. + 11.5404
    p['vz'] = (p['vr']*sinb + v_b*cosb) + 7.8146
    
    p.add_column(Column([rv_err]*len(dmu), name="err_vr"))
    p.add_column(Column(d_percent_err*p['dist'], name="err_dist"))
    p.add_column(Column(dmu, name="err_mul"))
    p.add_column(Column(dmu, name="err_mub"))
    
    return p
```

```python
subset_w_err_gaia_spitzer = uncertainties(subset, 0.02, 10.)
ascii.write(subset_w_err_gaia_spitzer, output='../../data/gaia_challenge/subsample_gaia_spitzer.txt', Writer=ascii.Basic, delimiter=',')
```

```python
subset_w_err_only_gaia = uncertainties(subset, 0.15, 10)
ascii.write(subset_w_err_only_gaia, output='../../data/gaia_challenge/subsample_pure_gaia.txt', Writer=ascii.Basic, delimiter=',')
```
