## Selecting from nearby wraps

The plan is to:

* Select all LM10 particles for one of the nearby wraps
* Bin by $\Lambda$, interpolate curves in $\Lambda$ vs. distance and $\Lambda$ vs. $B$
* Find all catalog stars with some distance from the curve

```python
def plot_wrap(wrap):
    fig,axes = plt.subplots(1,2,figsize=(10,6))
    axes[0].scatter(wrap['dist'], wrap['Lambda'],
               edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
    axes[0].set_xlim(0,70)
    axes[0].set_ylabel("$\Lambda_{sgr}$")

    axes[1].scatter(wrap['Beta'], wrap['Lambda'],
               edgecolor='none', color='k', marker='o', s=8, alpha=0.25)
    axes[1].set_ylabel("$\Lambda_{sgr}$")
    axes[1].set_xlabel("$B_{sgr}$")
    
    return fig, axes
```

```python
south_wrap = lm10_particle_data(expr="(Pcol>-1) & (Lmflag==1) & (Pcol<7) & "
                                     "(Lambda < 240) & (Lambda > 60)")
north_wrap = lm10_particle_data(expr="(Pcol>-1) & (Lmflag==-1) & (Pcol<7) & "
                                     "(Lambda > 200) & (Lambda < 340)")
```

```python
plot_wrap(south_wrap)
plot_wrap(north_wrap)
```

Bin and average in $\Lambda$, $B$, $D$

```python
from scipy.interpolate import interp1d
```

```python
def wrap_binner(wrap, Nbins=100):
    lambda_bins = np.linspace(wrap['Lambda'].min(), wrap['Lambda'].max(), Nbins)
    
    beta_per_bin = np.zeros_like(lambda_bins)
    dist_per_bin = np.zeros_like(lambda_bins)
    for ii,binL in enumerate(lambda_bins):
        if ii == Nbins-1: break
        
        binR = lambda_bins[ii+1]
        ps = wrap[(wrap['Lambda'] > binL) & (wrap['Lambda'] < binR)]
        beta_per_bin[ii] = np.median(ps['Beta'])
        dist_per_bin[ii] = np.median(ps['dist'])
    
    beta_func = interp1d(lambda_bins, beta_per_bin, kind='cubic')
    dist_func = interp1d(lambda_bins, dist_per_bin, kind='cubic')
    
    return beta_func, dist_func
```

```python
south_wrap.sort("Lambda")
fig,axes = plot_wrap(south_wrap)

S_beta_func, S_dist_func = wrap_binner(south_wrap, Nbins=50)
axes[0].plot(S_dist_func(south_wrap['Lambda']), south_wrap['Lambda'],
             linestyle='-', marker=None, lw=2, color='r')
axes[1].plot(S_beta_func(south_wrap['Lambda']), south_wrap['Lambda'],
             linestyle='-', marker=None, lw=2, color='r')
```

### Plot Radial Velocity vs. Lambda

```python
def wrap_to_RV(wrap):
    r = np.zeros((len(wrap), 3))
    v = np.zeros((len(wrap), 3))

    r[:,0] = wrap['x']
    r[:,1] = wrap['y']
    r[:,2] = wrap['z']

    v[:,0] = wrap['vx']
    v[:,1] = wrap['vy']
    v[:,2] = wrap['vz']
    
    return radial_velocity(r, v)
```

Merge the SDSS data into our original table

```python
wrap_candidates = catalog[idx]
len(wrap_candidates)

rv, rv_err = [], []
mu_l, mu_b = [], []
mu_err = []
for ii,star in enumerate(wrap_candidates):
    name = "a{0}".format(ii)
    
    if name in result['col0']:
        row = result[result['col0'] == name]
        if not np.isnan(star["vgsr"]):
            rv = gsr_to_hel(star["l"]*u.degree, 
                        star["b"]*u.degree, 
                        star["vgsr"]*u.km/u.s)
            rv_err = star["vgsr_err"]
        else:
            rv.append(float(row['elodiervfinal']))
            rv_err.append(float(row['elodiervfinalerr']))
        mu_l.append(float(row['pmL']))
        mu_b.append(float(row['pmB']))
        mu_err.append(float(np.sqrt(row['pmRaErr']**2 + row['pmDecErr']**2)))
    elif not np.isnan(star["vgsr"]):
        rv = gsr_to_hel(star["l"]*u.degree, 
                        star["b"]*u.degree, 
                        star["vgsr"]*u.km/u.s)
        rv_err = star["vgsr_err"]
        mu_l.append(np.nan)
        mu_b.append(np.nan)
        mu_err.append(np.nan)
    else:
        rv.append(np.nan)
        rv_err.append(np.nan)
        mu_l.append(np.nan)
        mu_b.append(np.nan)
        mu_err.append(np.nan)

c = Column(rv, name='rv')
wrap_candidates.add_column(c)
c = Column(rv_err, name='rv_err')
wrap_candidates.add_column(c)
c = Column(mu_l, name='mu_l')
wrap_candidates.add_column(c)
c = Column(mu_b, name='mu_b')
wrap_candidates.add_column(c)
c = Column(mu_err, name='mu_err')
wrap_candidates.add_column(c)
```
