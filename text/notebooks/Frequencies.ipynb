```python
import os
from streams.io.sgr import SgrSimulation
from streams.integrate import LeapfrogIntegrator
from streams.potential.lm10 import LawMajewski2010
from scipy.signal import lombscargle
```

```python
_m = "2.5e8"
m = float(_m)
sgr = SgrSimulation(_m)
s = sgr.satellite()
```

```python
lm10_potential = LawMajewski2010()
```

```python
N = 100
```

```python
X = np.random.uniform(-50, 50, size=(N,3))
V = (np.random.uniform(-300, 300, size=(N,3))*u.km/u.s).to(u.kpc/u.Myr).value
integrator = LeapfrogIntegrator(lm10_potential._acceleration_at,
                                X, V,
                                args=(X.shape[0], np.zeros_like(X)))
ts, rs, vs = integrator.run(t1=sgr.t1, t2=-10000, dt=-1.)
```

```python
fig,axes = subplots(1,2,figsize=(18,9),sharex=True,sharey=True)
for jj in range(N):
    axes[0].plot(rs[:,jj,0], rs[:,jj,2], alpha=0.5)
    axes[1].plot(rs[:,jj,1], rs[:,jj,2], alpha=0.5)
    xlim(-250,250)
    ylim(-250,250)
```

```python
#f = np.linspace(10, 10000, 250)
f = np.logspace(-4, -2, 250)

freqs = np.zeros((N,3))
for ii in range(3):
    for jj in range(N):
        pgram = lombscargle(ts, rs[:,jj,ii], f)
        freqs[jj,ii] = f[pgram.argmax()]
        #plot(f, np.sqrt(4*(pgram/rs.shape[0])))
```

```python
plot(freqs[:,1] / freqs[:,0],
     freqs[:,2] / freqs[:,0],
     marker='o', alpha=0.5, linestyle='none')

xlim(0.2, 1.5)
ylim(0.2, 1.5)
```
