```python
def run_back_integration(halo_potential, sgr_snap, sgr_cen):
    """ Given the particle snapshot information and a potential, integrate the particles
        backwards and return the minimum energy distances.
    """

    # Get timestep information from SGR_CEN
    t1 = min(sgr_cen.data["t"])
    t2 = max(sgr_cen.data["t"])
    dt = sgr_cen.data["dt"][0]

    # We use the same disk and bulge potentials for all runs, just vary the halo potential
    disk_potential = MiyamotoNagaiPotential(M=1E11*u.M_sun,
                                            a=6.5,
                                            b=0.26)
    bulge_potential = HernquistPotential(M=3.4E10*u.M_sun,
                                         c=0.7)
    potential = disk_potential + bulge_potential + halo_potential

    # Initialize particle simulation with full potential
    simulation = ParticleSimulation(potential=potential)

    for ii in range(sgr_snap.num):
        p = Particle(position=(sgr_snap.data["x"][ii], sgr_snap.data["y"][ii], sgr_snap.data["z"][ii]), # kpc
                     velocity=(sgr_snap.data["vx"][ii], sgr_snap.data["vy"][ii], sgr_snap.data["vz"][ii]), # kpc/Myr
                     mass=1.) # M_sol
        simulation.add_particle(p)

    # The data in SGR_CEN is only printed every 25 steps!
    ts, xs, vs = simulation.run(t1=t2, t2=t1, dt=-dt)
    ts = ts[::-25]
    xs = xs[::-25,:,:]
    vs = vs[::-25,:,:]

    # Define tidal radius, escape velocity for satellite
    msat = 2.5E8 # M_sun
    sgr_orbital_radius = np.sqrt(sgr_cen.data["x"]**2 + sgr_cen.data["y"]**2 + sgr_cen.data["z"]**2)
    m_halo_enclosed = halo_potential.params["v_halo"]**2 * sgr_orbital_radius/bulge_potential.params["_G"]
    mass_enclosed = disk_potential.params["M"] + bulge_potential.params["M"] + m_halo_enclosed

    r_tides = sgr_orbital_radius * (msat / mass_enclosed)**(1./3)
    v_escs = np.sqrt(bulge_potential.params["_G"] * msat / r_tides)

    closest_distances = []
    for ii in range(sgr_snap.num):
        # Distance to satellite center and total velocity
        d = np.sqrt((xs[:,ii,0] - sgr_cen.data["x"])**2 +
                    (xs[:,ii,1] - sgr_cen.data["y"])**2 +
                    (xs[:,ii,2] - sgr_cen.data["z"])**2)
        v = np.sqrt((vs[:,ii,0] - sgr_cen.data["vx"])**2 +
                    (vs[:,ii,1] - sgr_cen.data["vy"])**2 +
                    (vs[:,ii,2] - sgr_cen.data["vz"])**2)

        energy_distances = np.sqrt((d/r_tides)**2 + (v/v_escs)**2)
        closest_distances.append(min(energy_distances))

    return np.array(closest_distances)
```

Meeting with Hogg, DFM, Kathryn:

- Hogg suggests that if our errors are small relative to the thickness of the stream, $$p(x|x_{true}) \approx p(x_{true}|x)$$ because the prior over the true position in phase space (the width of the stream) is approximately flat over the likelihood of the data.

- Another idea was, instead of minimizing the median of the phase space distance distribution, minimize:

$$
\begin{equation}
\mathrm{Var}\Big(\frac{\vec{x}-\vec{x_{sgr}}}{R_{tide}}\Big) + \mathrm{Var}\Big(\frac{\vec{v}-\vec{v_{sgr}}}{v_{esc}}\Big) \sim \mathrm{Tr}(?)
\end{equation}
$$

```python
def hogg_run_back_integration(halo_potential, sgr_snap, sgr_cen):
    """ Given the particle snapshot information and a potential, integrate 
        the particles backwards.
    """

    # Get timestep information from SGR_CEN
    t1 = min(sgr_cen.data["t"])
    t2 = max(sgr_cen.data["t"])
    dt = sgr_cen.data["dt"][0]

    # We use the same disk and bulge potentials for all runs, just vary the halo potential
    disk_potential = MiyamotoNagaiPotential(M=1E11*u.M_sun,
                                            a=6.5,
                                            b=0.26)
    bulge_potential = HernquistPotential(M=3.4E10*u.M_sun,
                                         c=0.7)
    potential = disk_potential + bulge_potential + halo_potential

    # Initialize particle simulation with full potential
    simulation = ParticleSimulation(potential=potential)

    for ii in range(sgr_snap.num):
        p = Particle(position=(sgr_snap.data["x"][ii], sgr_snap.data["y"][ii], sgr_snap.data["z"][ii]), # kpc
                     velocity=(sgr_snap.data["vx"][ii], sgr_snap.data["vy"][ii], sgr_snap.data["vz"][ii]), # kpc/Myr
                     mass=1.) # M_sol
        simulation.add_particle(p)

    # The data in SGR_CEN is only printed every 25 steps!
    ts, xs, vs = simulation.run(t1=t2, t2=t1, dt=-dt)
    ts = ts[::-25]
    xs = xs[::-25,:,:]
    vs = vs[::-25,:,:]

    # Define tidal radius, escape velocity for satellite
    msat = 2.5E8 # M_sun
    sgr_orbital_radius = np.sqrt(sgr_cen.data["x"]**2 + sgr_cen.data["y"]**2 + sgr_cen.data["z"]**2)
    m_halo_enclosed = halo_potential.params["v_halo"]**2 * sgr_orbital_radius/bulge_potential.params["_G"]
    mass_enclosed = disk_potential.params["M"] + bulge_potential.params["M"] + m_halo_enclosed

    r_tides = sgr_orbital_radius * (msat / mass_enclosed)**(1./3)
    v_escs = np.sqrt(bulge_potential.params["_G"] * msat / r_tides)

    # Distance to satellite center and total velocity
    d = np.min(np.sqrt((xs[:,:,0] - sgr_cen.data["x"][:,np.newaxis].repeat(1000, axis=1))**2 +
                       (xs[:,:,1] - sgr_cen.data["y"][:,np.newaxis].repeat(1000, axis=1))**2 + 
                       (xs[:,:,2] - sgr_cen.data["z"][:,np.newaxis].repeat(1000, axis=1))**2) / r_tides[:,np.newaxis].repeat(1000, axis=1), axis=1)
    
    v = np.min(np.sqrt((vs[:,:,0] - sgr_cen.data["vx"][:,np.newaxis].repeat(1000, axis=1))**2 +
                       (vs[:,:,1] - sgr_cen.data["vy"][:,np.newaxis].repeat(1000, axis=1))**2 +
                       (vs[:,:,2] - sgr_cen.data["vz"][:,np.newaxis].repeat(1000, axis=1))**2) / v_escs[:,np.newaxis].repeat(1000, axis=1), axis=1)
    
    energy_trace = np.var(d) + np.var(v)
    return energy_trace
```

```python
def run_back_integration(halo_potential, sgr_snap, sgr_cen, dt):
    """ Given the particle snapshot information and a potential, integrate the particles
        backwards and return the minimum energy distances.
    """
    
    t1 = min(sgr_cen.data["t"])
    t2 = max(sgr_cen.data["t"])

    # We use the same disk and bulge potentials for all runs, just vary the halo potential
    disk_potential = MiyamotoNagaiPotential(M=1E11*u.M_sun,
                                            a=6.5,
                                            b=0.26)
    bulge_potential = HernquistPotential(M=3.4E10*u.M_sun,
                                         c=0.7)
    potential = disk_potential + bulge_potential + halo_potential

    # Initialize particle simulation with full potential
    simulation = ParticleSimulation(potential=potential)

    for ii in range(sgr_snap.num):
        p = Particle(position=(sgr_snap.data["x"][ii], sgr_snap.data["y"][ii], sgr_snap.data["z"][ii]), # kpc
                     velocity=(sgr_snap.data["vx"][ii], sgr_snap.data["vy"][ii], sgr_snap.data["vz"][ii]), # kpc/Myr
                     mass=1.) # M_sol
        simulation.add_particle(p)

    # The data in SGR_CEN is only printed every 25 steps!
    ts, xs, vs = simulation.run(t1=t2, t2=t1, dt=-dt)

    # Define tidal radius, escape velocity for satellite
    msat = 2.5E8 # M_sun
    sgr_orbital_radius = np.sqrt(sgr_cen.x**2 + sgr_cen.y**2 + sgr_cen.z**2)
    m_halo_enclosed = halo_potential.params["v_halo"]**2 * sgr_orbital_radius/bulge_potential.params["_G"]
    mass_enclosed = disk_potential.params["M"] + bulge_potential.params["M"] + m_halo_enclosed

    r_tides = sgr_orbital_radius * (msat / mass_enclosed)**(1./3)
    v_escs = np.sqrt(bulge_potential.params["_G"] * msat / r_tides)
    
    # N = number of particles, M = number of timesteps
    N = sgr_snap.num
    M = len(ts)
    
    # Distance to satellite center and total velocity
    d = np.sqrt((xs[:,:,0] - sgr_cen.x[:,np.newaxis].repeat(N, axis=1))**2 +
                (xs[:,:,1] - sgr_cen.y[:,np.newaxis].repeat(N, axis=1))**2 + 
                (xs[:,:,2] - sgr_cen.z[:,np.newaxis].repeat(N, axis=1))**2) / r_tides[:,np.newaxis].repeat(N, axis=1)
    
    v = np.sqrt((vs[:,:,0] - sgr_cen.vx[:,np.newaxis].repeat(N, axis=1))**2 +
                (vs[:,:,1] - sgr_cen.vy[:,np.newaxis].repeat(N, axis=1))**2 +
                (vs[:,:,2] - sgr_cen.vz[:,np.newaxis].repeat(N, axis=1))**2) / v_escs[:,np.newaxis].repeat(N, axis=1)
    
    idx = np.argmin(d**2 + v**2, axis=0)
    
    min_ds = []
    for ii,jj in enumerate(idx):
        min_ds.append(d[jj,ii])
    
    min_vs = []
    for ii,jj in enumerate(idx):
        min_vs.append(v[jj, ii])
    
    #print(np.var(min_ds) + np.var(min_vs))
    #print(np.var(np.min(np.sqrt(d**2 + v**2), axis=0)))
    
    return np.var(min_ds) + np.var(min_vs)
```

```python
import astropy.units as u
from streams.data import SgrSnapshot, SgrCen, read_lm10
from streams.data.gaia import add_uncertainties_to_particles
from streams.potential import LawMajewski2010
from streams.potential.lm10 import halo_params as true_halo_params
from streams.simulation import minimum_distance_matrix
```

```python
satellite_ic, particles = read_lm10(N=1000, expr="(Pcol > -1) & (abs(Lmflag) == 1) & (dist < 80)")
```

```python
t = np.arange(satellite_ic.t1,
                      satellite_ic.t2,
                      -5.)
t *= u.Myr
```

```python
true_halo_params
```

```python
particles = add_uncertainties_to_particles(particles, 
                                           radial_velocity_error=None,
                                           distance_error_percent=None)

# True potential
true_potential = LawMajewski2010(**true_halo_params)
particle_orbits = particles.integrate(true_potential, t)
satellite_orbit = satellite_ic.integrate(true_potential, t)
correct_ps = minimum_distance_matrix(true_potential, 
                                     particle_orbits, 
                                     satellite_orbit)

# Wrong potential
halo_params = true_halo_params.copy()
halo_params['qz'] *= 1.5
halo_params['q1'] *= 1.5
halo_params['v_halo'] *= 1.5
halo_params['phi'] *= 1.5
wrong_potential = LawMajewski2010(**halo_params)
particle_orbits = particles.integrate(wrong_potential, t)
satellite_orbit = satellite_ic.integrate(wrong_potential, t)
wrong_ps = minimum_distance_matrix(wrong_potential, 
                                   particle_orbits, 
                                   satellite_orbit)
```
