## Playing with Kathryn's simulation data of an Orphan-like stream

```python
import astropy.units as u
from astropy.constants import G
from astropy.io import ascii
```

```python
X = (G.decompose(bases=[u.kpc,u.M_sun,u.Myr]).value / 0.15**3 * 1E7)**-0.5
X
```

```python
length_unit = u.Unit("0.15 kpc")
mass_unit = u.Unit("1E7 M_sun")
time_unit = u.Unit("{:08f} Myr".format(X))

print length_unit, mass_unit, time_unit
```

```python
cen_colnames = ["t","dt","x","y","z","vx","vy","vz"]
cen_file = "/Users/adrian/projects/streams/data/simulation/Orphan/ORP_CEN"
orp_cen = ascii.read(cen_file, names=cen_colnames)
```

```python
(orp_cen["vz"]*length_unit/time_unit).to(u.km/u.s)
```

```python
subplot(221)
plot(orp_cen['x'], orp_cen['y'], linestyle='-', marker=None)
subplot(223)
plot(orp_cen['x'], orp_cen['z'], linestyle='-', marker=None)
subplot(224)
plot(orp_cen['y'], orp_cen['z'], linestyle='-', marker=None)
```
