```python
import astropy.units as u
from astropy.io import fits
import astropy.coordinates as coord
from astropy.table import Table, Column
from streams.coordinates.sgr import SgrCoordinates
from streams.coordinates import vgsr_to_vhel
```

```python
from streams.io.lm10 import particle_table
```

```python
lm10_particles = particle_table(N=0, expr="(Pcol > -1) & (Pcol < 7) & (abs(Lmflag) == 1)")
```

```python
hdulist = fits.open("../../data/catalog/apogee-dr10.fits")
```

```python
data = hdulist[1].data
data = Table(data)
```

```python
print data.colnames
```

```python
Lambda,Beta = [], []
for star in data:
    icrs = coord.ICRSCoordinates(star['RA'], star['DEC'], unit=(u.deg, u.deg))
    sgr = icrs.transform_to(SgrCoordinates)
    Lambda.append(sgr.Lambda.degrees)
    Beta.append(sgr.Beta.degrees)
```

```python
data.add_column(Column(Lambda, name='Lambda'))
data.add_column(Column(Beta, name='Beta'))
```
